### 8.3. Ph√¢n ph·ªëi poisson

Gi·∫£ s·ª≠ b·∫°n ƒëang ng·ªìi ·ªü m·ªôt b·ªánh vi·ªán. X√°c su·∫•t ƒë·ªÉ ph√≠a tr∆∞·ªõc b·∫°n c√≥ $x$ ng∆∞·ªùi l√† bao nhi√™u? Trung b√¨nh khi t·ªõi b·ªánh vi·ªán b·∫°n ph·∫£i ch·ªù trong th·ªùi gian bao l√¢u. Ph√¢n ph·ªëi poisson s·∫Ω c√≥ t√≠nh ·ª©ng d·ª•ng cao trong tr∆∞·ªùng h·ª£p n√†y. Ph√¢n ph·ªëi posson ƒë∆∞·ª£c ƒë·∫∑c tr∆∞ng b·ªüi hai tham s·ªë l√† $k$ ƒë·∫°i di·ªán cho s·ªë l∆∞·ª£t s·ª± ki·ªán x·∫£y ra v√† $\lambda$ l√† k·ª≥ v·ªçng c·ªßa ph√¢n ph·ªëi.

$$f(k; \lambda) = P(X=k) = \frac{\lambda^{k}e^{-\lambda}}{k!}$$

Ph√¢n ph·ªëi poisson tr√™n pytorch ƒë∆∞·ª£c kh·ªüi t·∫°o th√¥ng qua h√†m `torch.posson()`.

```python
rates = torch.rand(1000)
x = torch.poisson(rates)
plt.figure(figsize=(10, 6))
sns.kdeplot(x)
plt.xlabel("X")
plt.title("Poisson Distribution")
```

V√≠ d·ª•: Th·ªëng k√™ cho th·∫•y ƒë·ªô d√†i h√†ng ch·ªù kh√°m b·ªánh ·ªü b·ªánh vi·ªán tu√¢n theo ph√¢n ph·ªëi poisson v·ªõi k·ª≥ v·ªçng trung b√¨nh l√† $\lambda = 6.5$. H·ªèi x√°c su·∫•t ƒë·ªÉ c√≥ 2 ng∆∞·ªùi ƒë·ª©ng tr∆∞·ªõc b·∫°n khi x·∫øp h√†ng ·ªü b·ªánh vi·ªán l√† bao nhi√™u ?

$$f(k=3; \lambda) = P(X = 3) = \frac{\lambda^{k}e^{-\lambda}}{k!} = \frac{6.5^3e^{-6.5}}{3!}=0.069$$

Ch√∫ng ta c≈©ng c√≥ th·ªÉ t√≠nh x√°c su·∫•t cho ƒë·ªô d√†i h√†ng ch·ªù tu·ª≥ √Ω.

```python
import math
6.5**3*math.e**-6.5/6
```

```python
import numpy as np

def _poisson(k , lamb):
  arr = torch.arange(1, k+1)
  prod = 1
  for i in arr:
    prod*=i
  poss = lamb**k*np.e**(-lamb)/prod
  return poss

# L·∫≠p ph√¢n ph·ªëi posson v·ªõi lambda = 6.5 v√† ƒë·ªô d√†i h√†ng ch·ªù t·ª´ 1 t·ªõi 20.
lamb=6.5
poss = []
for k in torch.arange(1, 21):
  poss_i = _poisson(k, lamb)
  poss.append(poss_i)

for k, p in enumerate(poss):
    print("X√°c su·∫•t chi·ªÅu d√†i h√†ng ƒë·ª£i l√† {}: {:.2f}%".format(k+1, round(p.numpy()*100, 2)))

print("T·ªïng x√°c su·∫•t: {}".format(sum(poss)))
```

## 8.5. Ph√¢n ph·ªëi nh·ªã th·ª©c (_BINOMIAL DISTRIBUTION_)

Ph√¢n ph·ªëi nh·ªã th·ª©c cho ta bi·∫øt x√°c xu·∫•t ƒë·ªÉ 1 bi·∫øn c·ªë ki·ªán x·∫£y ra v·ªõi t·∫ßn su·∫•t l√† $k$ khi th·ª±c hi·ªán m·ªôt s·ªë l∆∞·ª£ng ph√©p th·ª≠ $n$, $n \geq k$. Ph√¢n ph·ªëi n√†y l√† tr∆∞·ªùng h·ª£p t·ªïng qu√°t c·ªßa ph√¢n ph·ªëi bernoulli cho t·∫ßn su·∫•t nhi·ªÅu h∆°n 1 bi·∫øn c·ªë x·∫£y ra v√† c√≥ h√†m ph√¢n ph·ªëi x√°c xu·∫•t l√†:
$$f_{p}(k) := P(X=k)=\binom{k}{n}p^{k}(1-p)^{n-k}$$

$\binom{k}{n}$ ch√≠nh l√† x√°c su·∫•t ƒë·ªÉ l·∫•y ng·∫´u nhi√™n t·ª´ $n$ m·∫´u ra $k$ ph·∫ßn t·ª≠ m√† kh√¥ng x√©t ƒë·∫øn t√≠nh th·ª© t·ª±. N√≥ c√≤n g·ªçi l√† t·ªï h·ª£p ch·∫≠p $k$ c·ªßa $n$.

$$\binom{k}{n} = \frac{n!}{k!(n-k)!}$$

V√≠ d·ª•: m·ªôt cu·ªôc kh·∫£o s√°t ƒë∆∞·ª£c th·ª±c hi·ªán ƒë·ªÉ t√¨m t·ª∑ l·ªá ng∆∞·ªùi th√≠ch m√®o h∆°n ch√≥ l√†m th√∫ c∆∞ng. C√≥ 100 ng∆∞·ªùi ƒë∆∞·ª£c kh·∫£o s√°t ng·∫´u nhi√™n, v√† ph·∫£n h·ªìi c·ªßa h·ªç ƒë∆∞·ª£c ghi nh·∫≠n. Gi·∫£ s·ª≠ r·∫±ng t·ª∑ l·ªá th·ª±c t·∫ø (true proportion) nh·ªØng ng∆∞·ªùi th√≠ch m√®o h∆°n ch√≥ trong t·ªïng kh·∫£o s√°t l√† 0.6.

T√¨nh hu·ªëng n√†y c√≥ th·ªÉ ƒë∆∞·ª£c m√¥ h√¨nh h√≥a b·∫±ng ph√¢n ph·ªëi nh·ªã th·ª©c. G·ªçi ùëã l√† s·ªë ng∆∞·ªùi trong m·∫´u kh·∫£o s√°t th√≠ch m√®o. Khi ƒë√≥, ùëã tu√¢n theo ph√¢n ph·ªëi nh·ªã th·ª©c v·ªõi:

n=100: K√≠ch th∆∞·ªõc m·∫´u (s·ªë ng∆∞·ªùi ƒë∆∞·ª£c kh·∫£o s√°t).
 
p=0.6: T·ª∑ l·ªá th·ª±c t·∫ø nh·ªØng ng∆∞·ªùi th√≠ch m√®o.

Y√™u c·∫ßu:

T√¨m x√°c su·∫•t ƒë·ªÉ c√≥ t·ªëi ƒëa 50 ng∆∞·ªùi trong m·∫´u kh·∫£o s√°t th√≠ch m√®o. X√°c su·∫•t n√†y c√≥ th·ªÉ ƒë∆∞·ª£c t√≠nh b·∫±ng h√†m ph√¢n ph·ªëi t√≠ch l≈©y (CDF) c·ªßa ph√¢n ph·ªëi nh·ªã th·ª©c:

```python
import numpy as np
from scipy.special import comb

# C√¥ng th·ª©c t√≠nh t·ªï h·ª£p
def _combine(k, n):
#  assert (k>=0) & (n>=k)
  return comb(n, k, exact=False)

# T√≠nh ph√¢n ph·ªëi nh·ªã th·ª©c
def _binary_dist(n, k, p):
#  assert (k>=0) & (n>=k)
  c_k_n = _combine(k, n)
  prob = c_k_n*p**k*(1-p)**(n-k)
  return prob

def cumulative_binomial(n,max_k, p):
    total_prob = 0
    for k in range(max_k + 1): 
      total_prob += _binary_dist(n, k, p)  # C·ªông d·ªìn x√°c su·∫•t
    return total_prob

# k = 50, n = 100, p = 0.6
k = 50
n = 100
p = 0.6
print("X√°c su·∫•t ph√¢n ph·ªëi nh·ªã ph√¢n v·ªõi k = 50, n = 100, p = 0.6:")
pro=cumulative_binomial(n, k, p)
print(pro)
```
