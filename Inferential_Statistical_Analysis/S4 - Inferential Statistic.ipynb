## 1. Giới thiệu

Thống kê suy diễn là một nhánh của thống kê, tập trung vào việc đưa ra suy luận hoặc kết luận về một population dựa trên thông tin thu thập được từ một sample. Phương pháp này sử dụng dữ liệu sample để rút ra suy luận về một population lớn hơn.

Thống kê suy diễn áp dụng lý thuyết xác suất để thực hiện các suy luận này. Nó giúp các nhà nghiên cứu đưa ra các tổng quát hóa về một population bằng cách nghiên cứu một tập hợp con các cá nhân hoặc đối tượng từ quần thể đó.

Để thực hiện thống kê suy diễn, thông thường bắt đầu bằng một giả thuyết hoặc tuyên bố về population. Tiếp theo, thu thập một mẫu dữ liệu từ population và sử dụng các phương pháp thống kê để phân tích dữ liệu mẫu. Dựa trên kết quả phân tích, đưa ra suy luận về population.

Thống kê suy diễn có thể được sử dụng trong nhiều lĩnh vực khác nhau, bao gồm nghiên cứu thị trường, nghiên cứu y khoa, khoa học xã hội, và nhiều lĩnh vực khác. Nó cho phép các nhà nghiên cứu rút ra kết luận về một population với mức độ tin cậy nhất định và hỗ trợ đưa ra các quyết định dựa trên dữ liệu

`Parameter và statistic parameter:`

- Parameters: Tham số là các đặc trưng số học mô tả một quần thể. Chúng thường được ký hiệu bằng các chữ cái Hy Lạp như µ (mu) hoặc σ (sigma). Tham số thường được sử dụng để suy luận về quần thể dựa trên một mẫu.

Ví dụ: Để biết chiều cao trung bình của tất cả mọi người trong một quốc gia, có thể lấy một mẫu người và tính trung bình mẫu. Trung bình mẫu là một ước lượng cho trung bình quần thể, chính là tham số mà ta quan tâm.

- Statistics estimates: Ước số thống kê là các đặc trưng số học mô tả cho mẫu. Chúng thường được ký hiệu bằng các chữ cái Latinh như x̄ (x-bar) hoặc s. Ước số được sử dụng để tóm tắt và phân tích dữ liệu mẫu, cũng như để suy luận về quần thể.

Ví dụ: Nếu ta lấy một mẫu người và tính trung bình mẫu, thì trung bình mẫu đó là một giá trị thống kê. Giá trị thống kê này có thể được sử dụng để suy luận về trung bình quần thể, là tham số mà ta quan tâm.

## 2. Sampling Theory

Lý thuyết lấy mẫu là một lĩnh vực trong thống kê, tập trung vào việc chọn các phần tử hoặc đối tượng từ một population lớn hơn. Mục tiêu của việc lấy mẫu là thu thập thông tin về population bằng cách nghiên cứu các đặc điểm của mẫu. Lý thuyết lấy mẫu cung cấp một khuôn khổ để chọn một mẫu đại diện cho population và ước lượng các tham số quan tâm, chẳng hạn như trung bình, phương sai hoặc tỷ lệ.

Trên thực tế, lý thuyết lấy mẫu được áp dụng trong nhiều lĩnh vực, như nghiên cứu thị trường, khoa học xã hội, y tế công cộng, và sản xuất. Ví dụ, một công ty nghiên cứu thị trường có thể sử dụng lý thuyết lấy mẫu để chọn một mẫu đại diện của người tiêu dùng nhằm khảo sát sở thích của họ đối với một sản phẩm mới. 

Trong y tế công cộng, các nhà nghiên cứu có thể sử dụng lý thuyết này để chọn một mẫu bệnh nhân từ quần thể để nghiên cứu tỷ lệ mắc một căn bệnh.

Lý thuyết lấy mẫu bao gồm nhiều kỹ thuật chọn mẫu khác nhau, như random sampling (lấy mẫu ngẫu nhiên), stratified sampling (lấy mẫu phân tầng), cluster sampling (lấy mẫu cụm), systematic sampling (lấy mẫu hệ thống). Việc lựa chọn kỹ thuật lấy mẫu phụ thuộc vào đặc điểm của population và mục tiêu nghiên cứu. Khi mẫu đã được chọn, các phương pháp thống kê có thể được sử dụng để ước lượng các tham số của population và định lượng độ chính xác cũng như độ tin cậy của các ước lượng đó.



`Phân loại:`

- Lấy mẫu ngẫu nhiên (Simple Random Sampling)

Trong phương pháp này, mỗi cá nhân trong quần thể đều có khả năng được chọn ngang nhau. Thông thường, phương pháp này gắn giá trị số cho từng thành viên và sử dụng bộ tạo số ngẫu nhiên để chọn.

Sử dụng khi có danh sách đầy đủ của toàn bộ quần thể và muốn đảm bảo rằng mỗi cá nhân đều có cơ hội ngang nhau được chọn.

Ví dụ: Một trường đại học muốn khảo sát ý kiến của sinh viên về chất lượng giảng dạy. Họ lập danh sách toàn bộ (ví dụ có 10.000) sinh viên và sử dụng phần mềm tạo số ngẫu nhiên để chọn ra 500 sinh viên tham gia khảo sát.

- Lấy mẫu phân tầng (Stratified Sampling)

Ở đây, quần thể được phân thành các nhóm nhỏ (hoặc tầng) dựa trên các đặc điểm cụ thể như công việc, tuổi, giới tính, hoặc thu nhập... Sau đó, các thành viên được chọn ngẫu nhiên từ từng nhóm nhỏ để tạo ra một mẫu đại diện chính xác cho toàn bộ quần thể.

Sử dụng khi quần thể có sự khác biệt rõ rệt giữa các nhóm (tầng) và đảm bảo rằng mỗi nhóm đều được đại diện trong mẫu

Ví dụ: Một công ty muốn khảo sát mức độ hài lòng của nhân viên dựa trên cấp bậc (nhân viên, quản lý, và giám đốc). Công ty chia nhân viên thành 3 nhóm theo cấp bậc và chọn ngẫu nhiên 50 người từ mỗi nhóm để đảm bảo mỗi cấp bậc đều được đại diện

- Lấy mẫu cụm (Cluster Sampling)

Lấy mẫu cụm bao gồm việc chia quần thể thành các cụm hoặc nhóm, sau đó chọn ngẫu nhiên một số cụm và lấy tất cả thành viên trong những cụm được chọn vào mẫu.

Sử dụng khi quần thể rất lớn và việc thu thập dữ liệu từ toàn bộ quần thể là không khả thi, nhưng có thể chia nhỏ thành các cụm dễ tiếp cận

Ví dụ: Một nhà nghiên cứu muốn khảo sát thói quen ăn uống của học sinh tại một thành phố. Họ chia các trường học trong thành phố thành các cụm và ngẫu nhiên chọn 5 trường. Sau đó, họ khảo sát tất cả học sinh trong 5 trường được chọn

- Lấy mẫu hệ thống (Systematic Sampling)

Phương pháp này sắp xếp quần thể theo một thứ tự hoặc danh sách. Một điểm bắt đầu ngẫu nhiên được chọn, sau đó chọn các thành viên thứ n trong danh sách để đưa vào mẫu.

Sử dụng khi có danh sách hoặc thứ tự sắp xếp của quần thể và muốn chọn mẫu một cách nhanh chóng, đơn giản mà vẫn đảm bảo tính ngẫu nhiên

Ví dụ: Một cửa hàng muốn khảo sát ý kiến khách hàng theo ngày. Họ sắp xếp danh sách khách hàng theo thứ tự thời gian mua hàng và chọn ngẫu nhiên một người làm điểm bắt đầu, sau đó chọn mỗi khách hàng thứ 50 trong danh sách để khảo sát.

- Lấy mẫu tiện lợi (Convenience Sampling)

Lấy mẫu tiện lợi chọn những người tham gia dễ dàng tiếp cận và sẵn sàng tham gia, chẳng hạn như học sinh trong một lớp học hoặc khách hàng trong một cửa hàng. Tuy nhiên, phương pháp này thường không đại diện cho toàn bộ quần thể.

Sử dụng khi cần thu thập dữ liệu một cách nhanh chóng, chi phí thấp, hoặc khi quần thể mục tiêu khó tiếp cận. 

Ví dụ: Một sinh viên muốn khảo sát ý kiến về việc sử dụng mạng xã hội. Họ chỉ phỏng vấn những người bạn thân hoặc những người dễ dàng tiếp cận trong lớp học của mình.

- Lấy mẫu quả cầu tuyết (Snowball Sampling)

Phương pháp này dựa vào việc những người tham gia giới thiệu và chỉ định những người khác. Phương pháp này thường được sử dụng khi quần thể mục tiêu khó xác định hoặc tiếp cận.

Sử dụng khi quần thể mục tiêu khó xác định hoặc khó tiếp cận, chẳng hạn như các nhóm nhỏ, cộng đồng bí mật, hoặc nhóm người có đặc điểm chung hiếm gặp

Ví dụ:Một nhà nghiên cứu muốn tìm hiểu về những người tham gia vào các cộng đồng riêng tư trên mạng xã hội. Họ bắt đầu với một vài người tham gia và yêu cầu họ giới thiệu thêm những người khác trong cộng đồng để tiếp tục mở rộng mẫu

`Áp dụng trong:`

- Nghiên cứu thị trường

Trong nghiên cứu thị trường, một mẫu người tiêu dùng có thể được chọn từ một quần thể lớn hơn để thu thập thông tin về sở thích, hành vi và thái độ của họ đối với một sản phẩm hoặc dịch vụ.

- Nghiên cứu y học

Trong nghiên cứu y học, một mẫu bệnh nhân có thể được chọn từ một quần thể lớn hơn để nghiên cứu hiệu quả của một phương pháp điều trị hoặc một loại thuốc.

- Thăm dò ý kiến

Trong thăm dò ý kiến, một mẫu cử tri có thể được chọn từ một quần thể lớn hơn để dự đoán kết quả của một cuộc bầu cử hoặc đo lường ý kiến công chúng về một vấn đề cụ thể.

- Kiểm soát chất lượng

Trong kiểm soát chất lượng, một mẫu sản phẩm có thể được chọn từ một lô hàng lớn hơn để kiểm tra lỗi hoặc đảm bảo rằng chúng đáp ứng các tiêu chuẩn chất lượng nhất định.

- Giám sát môi trường

Trong giám sát môi trường, một mẫu đất, nước hoặc không khí có thể được lấy từ một khu vực lớn hơn để xác định mức độ ô nhiễm hoặc chất gây ô nhiễm.

- Kiểm toán

Trong kiểm toán, một mẫu giao dịch tài chính có thể được chọn từ một tập dữ liệu lớn hơn để kiểm tra lỗi hoặc hành vi gian lận.

## 3. Phân phối mẫu

Trong thống kê, phân phối mẫu (sampling distribution) là phân phối của một thống kê thu được từ một số lượng lớn các mẫu được rút ra từ một quần thể. Nó mô tả sự phân bố của các giá trị của một thống kê (chẳng hạn như trung bình hoặc độ lệch chuẩn) nếu nhiều mẫu ngẫu nhiên được lấy từ quần thể và tính toán thống kê cho từng mẫu.

Phân phối mẫu rất quan trọng vì nó cho phép suy luận về quần thể dựa trên dữ liệu mẫu. Bằng cách xem xét phân phối mẫu của một thống kê, chúng ta có thể ước tính tham số của quần thể và xác định khả năng rằng một giá trị thống kê cụ thể được rút ra từ quần thể.

Ví dụ: Giả sử cần ước tính chiều cao trung bình của tất cả sinh viên trong một trường đại học. Một mẫu có thể được lấy, ví dụ 100 sinh viên, và tính toán chiều cao trung bình của họ. Tuy nhiên, trung bình mẫu này có thể không hoàn toàn bằng với trung bình thực tế của quần thể. Phân phối mẫu của trung bình cho biết khả năng trung bình mẫu của bạn gần với trung bình thực tế của quần thể.

Định lý giới hạn trung tâm (Central Limit Theorem) là một khái niệm quan trọng trong lý thuyết phân phối mẫu. Định lý này nêu rằng, dưới một số điều kiện nhất định, phân phối mẫu của một thống kê (chẳng hạn như trung bình) sẽ xấp xỉ phân phối chuẩn, bất kể phân phối của quần thể ban đầu. Điều này làm cho việc sử dụng thống kê suy diễn trở nên khả thi để đưa ra dự đoán về quần thể dựa trên mẫu.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

population = np.random.uniform(low=0, high=10, size=100000)  # phân phối đều
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Population")
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")
```

```python
def tb_mau(population, sample_size, num_samples):
    tb = []
    for i in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        tb.append(np.mean(sample)) 
    return tb
```

```python
sample_size = 30  # Kích thước mỗi mẫu
num_samples = 1000  # Số lượng mẫu được lấy
tb = tb_mau(population, sample_size, num_samples)


plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.title("Population")
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")

plt.subplot(1, 2, 2)
sns.histplot(tb, bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title("Population theo mean (Central limit theory)")
plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần suất")

plt.show()
```

## 4. Standard error

Sai số chuẩn là một thuật ngữ thống kê đo lường mức độ biến động của một thống kê mẫu, chẳng hạn như trung bình hoặc độ lệch chuẩn, so với tham số của quần thể. Nó chính là độ lệch chuẩn của phân phối mẫu của một thống kê, và là một ước lượng của tham số quần thể dựa trên một mẫu quan sát.

Nói cách khác, sai số chuẩn đo lường mức độ biến động hoặc không chắc chắn trong việc ước lượng tham số quần thể dựa trên một mẫu. Độ chính xác của ước lượng càng cao thì sai số chuẩn càng nhỏ. Ngược lại, khi sai số chuẩn lớn hơn, ước lượng có xu hướng kém chính xác hơn.

Công thức tính sai số chuẩn phụ thuộc vào loại thống kê đang được sử dụng. Dưới đây là các công thức cho ba loại thống kê phổ biến:

- SE của trung bình:

$$\text{SE} = \frac{\sigma}{\sqrt{n}}$$

    - σ là độ lệch chuẩn của mẫu.
    - n là kích thước mẫu

- SE của tỷ lệ:

$$\text{SE} = \sqrt{\frac{p(1-p)}{n}}$$

    - p là tỷ lệ mẫu.
    - n là kích thước mẫu

- SE của trung bình 2 mẫu:

$$\text{SE} = \sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}$$

    - σ1, σ1 là độ lệch chuẩn của 2 mẫu.
    - n1, n2 là kích thước của 2 mẫu

Ví dụ 1: có một mẫu gồm 20 học sinh và cần ước tính chiều cao trung bình của tất cả học sinh trong một trường học. Chiều cao của từng học sinh trong mẫu được đo và tính trung bình mẫu là 170 cm. Độ lệch chuẩn của mẫu là 5 cm.

Để tính sai số chuẩn của trung bình (SE), công thức sau được sử dụng:

$\text{SE} = \frac{\sigma}{\sqrt{n}}$
​
 
Trong đó:

σ là độ lệch chuẩn của mẫu.

n là kích thước mẫu.

Trong ví dụ này:

s=5cm

n=20

Tính toán:

𝑆𝐸 = 5/sqrt(20) = 5/4.472 = 1.118
 
Do đó, sai số chuẩn của trung bình là 1.118 cm.

Khi tính chiều cao trung bình cho nhiều mẫu có cùng kích thước được lấy từ cùng một quần thể, độ lệch chuẩn của các giá trị trung bình mẫu sẽ xấp xỉ 1.118 cm. Điều này có nghĩa là trung bình mẫu có thể dao động khoảng 1.118 đơn vị quanh trung bình thực của quần thể.

- Nếu SE nhỏ, trung bình mẫu gần như sẽ rất gần với trung bình thực của quần thể. Điều này cho thấy mẫu đại diện tốt cho quần thể.
- Nếu SE lớn, trung bình mẫu có thể dao động nhiều, làm giảm độ chính xác khi dự đoán trung bình quần thể.

Ví dụ 2: Giả sử trong một khảo sát, p=0.6 (60% người tham gia khảo sát đồng ý với một ý kiến), và kích thước mẫu là n=100. Khi đó:

SE=sqrt(0.6(1-0.6)/100) = 0.049

Điều này có nghĩa là tỷ lệ mẫu có thể dao động khoảng 4.9% quanh tỷ lệ thực của quần thể.

- SE nhỏ (SE =0.049) cho thấy tỷ lệ mẫu là một ước tính đáng tin cậy cho tỷ lệ thực của quần thể.
- SE lớn (SE =0.2) cho thấy tỷ lệ mẫu có độ không chắc chắn cao và không đại diện tốt cho quần thể.

Ví dụ 3: Giáo viên muốn so sánh điểm trung bình môn Toán của học sinh nam và nữ tại một trường trung học. Sau khi thu thập dữ liệu, họ có thông tin sau:

Nhóm 1 (Nam):
- Độ lệch chuẩn (σ1) = 8
- Kích thước mẫu (n1) = 50

Nhóm 2 (Nữ):
- Độ lệch chuẩn (σ2) = 10
- Kích thước mẫu (n2) = 40

$$\text{SE} = \sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}$$

Thế vào ta được SE=1.94

Sai khác giữa trung bình điểm Toán của học sinh nam và nữ có thể dao động khoảng 1.94 điểm quanh sai khác thực giữa hai trung bình của quần thể.

Nếu sai khác thực giữa trung bình điểm của hai nhóm là nhỏ (ví dụ: dưới 1.94), có thể kết luận rằng không có sự khác biệt đáng kể giữa nam và nữ về điểm Toán (dựa trên kích thước mẫu này và độ lệch chuẩn đã cho).

## 5. Hypothesis Testing

Kiểm định giả thuyết là một phương pháp thống kê được sử dụng để xác định liệu một giả thuyết về một tham số của quần thể có được giải thích bởi dữ liệu mẫu hay không. Quá trình này bao gồm các bước:

- Xây dựng giả thuyết về tham số của quần thể (thường gồm giả thuyết không H0 và giả thuyết thay thế Ha).
- Thu thập dữ liệu mẫu từ quần thể.
- Phân tích dữ liệu bằng các phương pháp thống kê để kiểm tra tính hợp lệ của giả thuyết.
- Kết luận về giả thuyết, dựa trên kết quả phân tích dữ liệu và mức ý nghĩa đã chọn.

Phương pháp này giúp đánh giá mức độ tin cậy của giả thuyết dựa trên dữ liệu thực nghiệm

### 5.1. Giả thuyết không H0 (null) và giả thuyết thay thế Ha

### a. Giả thuyết không (H0): 

- Là phát biểu mặc định được giả định là đúng cho đến khi có đủ bằng chứng để bác bỏ.

- Thường mô tả tình huống "không có sự khác biệt", "không có tác động", hoặc "giá trị quan sát được không đáng kể".

- Mục tiêu của kiểm định giả thuyết là xác định xem có nên bác bỏ H0 hay không, dựa trên dữ liệu mẫu.

Ví dụ:

Giả thuyết không: "Không có sự khác biệt giữa chiều cao trung bình của nam và nữ."

Giả thuyết không: "Thuốc mới không có tác dụng khác biệt so với thuốc cũ."

### b. Giả thuyết thay thế (Ha)

- Là phát biểu mâu thuẫn với H0, thường là điều mà nhà nghiên cứu mong muốn chứng minh.

- Thường biểu thị sự khác biệt hoặc một mối quan hệ giữa các biến.

- Nếu dữ liệu mẫu cung cấp bằng chứng mạnh mẽ, H0 bị bác bỏ và Ha được chấp nhận.

Ví dụ:

Giả thuyết thay thế: "Chiều cao trung bình của nam khác chiều cao trung bình của nữ."

Giả thuyết thay thế: "Thuốc mới có tác dụng khác biệt so với thuốc cũ."

### c. Tính chất: 

- Mối quan hệ giữa H0 và Ha là hai khả năng đối lập, không thể xảy ra đồng thời. Trong kiểm định, người ta không chứng minh trực tiếp Ha
​đúng mà sẽ bác bỏ H0 khi có bằng chứng thống kê đủ mạnh.

- Nếu không bác bỏ H0: Dữ liệu không đủ bằng chứng để chấp nhận Ha.

- Nếu bác bỏ H0: Có đủ bằng chứng để chấp nhận Ha.

Ví dụ 1: So sánh điểm số giữa hai nhóm học sinh

Giả thuyết không (H0): Không có sự khác biệt về điểm số giữa học sinh học 2 giờ và học sinh học 4 giờ. Điểm số trung bình của hai nhóm là như nhau, hoặc sự khác biệt (nếu có) chỉ là ngẫu nhiên.

Giả thuyết thay thế (Ha): Học sinh học 4 giờ sẽ đạt điểm cao hơn so với học sinh học 2 giờ. Có sự khác biệt thực sự về điểm số, và thời gian học tập lâu hơn giúp cải thiện điểm.

Ví dụ 2: Quan hệ giữa tiêu thụ caffeine và nhịp tim

Giả thuyết không (H0): Không có mối quan hệ nào giữa việc tiêu thụ caffeine và nhịp tim. Tiêu thụ caffeine không ảnh hưởng đến nhịp tim, hoặc sự thay đổi (nếu có) chỉ là ngẫu nhiên.

Giả thuyết thay thế (Ha): Việc tăng tiêu thụ caffeine dẫn đến tăng nhịp tim. Có mối quan hệ thực sự, và caffeine ảnh hưởng trực tiếp đến việc tăng nhịp tim

### 5.2. Các bước kiểm định giả thuyết

### a. Phát biểu giả thuyết

- Giả thuyết không (H0): Đây là giả định mặc định được kiểm định, thường thể hiện "không có sự khác biệt" hoặc "không có mối quan hệ".

- Giả thuyết thay thế (Ha): Đây là điều mà người thống kê muốn chứng minh, thể hiện "có sự khác biệt" hoặc "có mối quan hệ".

Ví dụ:

H0: "Không có sự khác biệt về điểm trung bình giữa hai nhóm học sinh."

Ha: "Điểm trung bình của nhóm học 4 giờ cao hơn nhóm học 2 giờ."

### b. Chọn mức ý nghĩa (α)

- Mức ý nghĩa là xác suất tối đa mà ta chấp nhận mắc sai lầm loại I (bác bỏ H0 khi H0 đúng). 

- Thông thường: chọn α=0.05 hoặc α=0.01, tương ứng với mức độ tin cậy 95% hoặc 99%.

### c. Thu thập dữ liệu và tính toán thống kê kiểm định

- Thu thập mẫu dữ liệu cần thiết.

- Sử dụng công thức hoặc phần mềm thống kê để tính toán giá trị thống kê kiểm định (test statistic), như t-value, z-value, hoặc F-value, tùy thuộc vào loại kiểm định (t-test, z-test, ANOVA, ...).

Ví dụ:

Nếu so sánh hai trung bình mẫu: Dùng t-test.

Nếu kiểm tra tỷ lệ: Dùng z-test.

### d. So sánh giá trị thống kê kiểm định với giá trị tới hạn (critical value) hoặc p-value

- Sử dụng giá trị tới hạn (critical): Dựa vào bảng phân phối (như z-table hoặc t-table) và mức ý nghĩa α, xác định giá trị tới hạn.

Nếu `giá trị thống kê kiểm định lớn hơn giá trị tới hạn`, bác bỏ H0.

- Sử dụng p-value: Tính p-value từ thống kê kiểm định.

Nếu `p-value < α`, bác bỏ H0.

Kết luận

- Nếu bác bỏ H0: Có đủ bằng chứng để chấp nhận Ha.

- Nếu không bác bỏ H0: Không đủ bằng chứng để bác bỏ H0, nhưng điều này không chứng minh rằng H0 đúng.

### 5.3. Lỗi kiểm định

### a. Lỗi loại 1 (Type 1 Error)

Trong kiểm định giả thuyết thống kê, lỗi loại 1 (còn gọi là dương tính giả) xảy ra khi bác bỏ giả thuyết không (H0) mặc dù nó thực sự đúng. Nói cách khác, kết luận rằng có sự khác biệt hoặc tác động khi thực tế không có.

Ký hiệu: α (alpha).

Thường được đặt ở mức ý nghĩa 0.05 (5%) hoặc 0.01 (1%), tức xác suất xảy ra lỗi loại 1 là 5% hoặc 1%.

Cách giảm lỗi loại 1:

    - Tăng kích thước mẫu.
    - Giảm mức ý nghĩa (α) từ 0.05 xuống 0.01.
    - Sử dụng các phương pháp thống kê nghiêm ngặt hơn.

### b. Lỗi loại 2 (Type 2 Error)

Lỗi loại 2 (còn gọi là âm tính giả) xảy ra khi không bác bỏ giả thuyết không (H0) mặc dù nó thực sự sai. Nói cách khác, kết luận rằng không có sự khác biệt hoặc tác động khi thực tế có.

Ký hiệu: β (beta).

Cách giảm lỗi loại 2:

    - Tăng kích thước mẫu.
    - Sử dụng các kiểm định thống kê nhạy hơn, có lực kiểm định (power) cao hơn.
    - Tuy nhiên, việc giảm nguy cơ mắc lỗi loại 2 thường đi kèm với việc tăng nguy cơ mắc lỗi loại 1, nên phải chấp nhận tăng nhẹ xác suất xảy ra lỗi loại 1 (α).

## 6. Critical region

Trong kiểm định giả thuyết thống kê, vùng tới hạn là tập hợp các giá trị dẫn đến việc bác bỏ giả thuyết không. 

Trong đó, mức ý nghĩa, ký hiệu là alpha (α), là xác suất tối đa có thể xảy ra của lỗi loại 1, tức là bác bỏ một giả thuyết không đúng.

Mức ý nghĩa xác định vùng tới hạn. Ví dụ, nếu mức ý nghĩa được đặt là 0.05 (hoặc 5%), có thể chấp nhận 5% khả năng xảy ra lỗi loại 1. Do đó, vùng tới hạn sẽ là tập hợp các giá trị rơi vào 5% cực trị của phân phối, ở các phần đuôi của phân phối.

Vùng tới hạn được xác định dựa trên giả thuyết không, giả thuyết thay thế, và mức ý nghĩa đã chọn. Nếu thống kê kiểm định rơi vào vùng tới hạn, giả thuyết không sẽ bị bác bỏ, và kết luận rằng có bằng chứng hỗ trợ cho giả thuyết thay thế. Nếu thống kê kiểm định không nằm trong vùng tới hạn, không thể bác bỏ giả thuyết không.



Với mức ý nghĩa α=0.05:
- Trong kiểm định một phía, vùng tới hạn chiếm 5% ở một phía của phân phối.
- Trong kiểm định hai phía, vùng tới hạn chiếm 2.5% ở mỗi phía của phân phối.

### 7. Confidence interval

Khoảng tin cậy (Confidence Interval - CI) là một khoảng giá trị được tính toán từ dữ liệu mẫu, dùng để ước lượng giá trị thực của một tham số trong tổng thể, chẳng hạn như trung bình hoặc tỷ lệ.

Nó cho biết phạm vi mà tham số tổng thể có khả năng nằm trong đó với một mức độ tin cậy nhất định, thường được biểu diễn dưới dạng phần trăm, như 95% hoặc 99%.

Ví dụ:
Cần tính khoảng tin cậy 95% cho trọng lượng trung bình của một quần thể chó:

Kích thước mẫu (n) = 50.
Trọng lượng trung bình mẫu (X) = 25 kg.
Độ lệch chuẩn mẫu (σ) = 3 kg.
Bước 1: Tính SEM:

SEM= σ/sqrt(n)=3/sqrt(50) =0.424

Bước 2: Xác định giá trị tới hạn:

Với khoảng tin cậy 95%, giá trị tới hạn là 1.96 (tra bảng phân phối chuẩn hoặc sử dụng máy tính).

Bước 3: Tính khoảng tin cậy:


CI=X±(giá trị tới hạn×SEM)

CI=25±(1.96×0.424)=[24.17,25.83]

Dựa trên dữ liệu mẫu, ta có 95% tin tưởng rằng trọng lượng trung bình thực sự của quần thể chó nằm trong khoảng từ 24.17 kg đến 25.83 kg.

### 8. One tail, two tail test

- Kiểm định một phía (One-tailed test), còn được gọi là kiểm định có hướng, là một phương pháp kiểm định giả thuyết thống kê. Trong đó, người thống kê hoặc phân tích đưa ra dự đoán cụ thể về hướng của mối quan hệ giữa hai biến. Ở kiểm định một phía, giả thuyết không bị bác bỏ nếu thống kê mẫu rơi vào một trong hai phía của phân phối mẫu: phía trên hoặc phía dưới, tùy thuộc vào hướng của giả thuyết.

Ví dụ:

Giả sử một nhà nghiên cứu muốn kiểm định giả thuyết rằng một loại thuốc mới sẽ tăng tuổi thọ trung bình của bệnh nhân:

Nếu dự đoán rằng thuốc sẽ làm tăng tuổi thọ, kiểm định một phía sẽ là kiểm định phía phải. Giả thuyết không sẽ là "thuốc không có tác dụng."
Trong trường hợp này, giả thuyết không bị bác bỏ nếu trung bình mẫu của tuổi thọ cao hơn đáng kể so với trung bình kỳ vọng.

Ngược lại, nếu nhà nghiên cứu dự đoán rằng thuốc sẽ làm giảm tuổi thọ, kiểm định một phía sẽ là kiểm định phía trái.Trong trường hợp này, giả thuyết không sẽ bị bác bỏ nếu trung bình mẫu của tuổi thọ thấp hơn đáng kể so với trung bình kỳ vọng.

- Kiểm định một phía dưới (Lower-Tailed Test): Giả thuyết thay thế là 𝐻1<𝜇0



- Kiểm định một phía trên (Upper-Tailed Test): Giả thuyết thay thế là 𝐻1>𝜇0




- Kiểm định hai phía (Two-tailed test) là một phương pháp kiểm định giả thuyết thống kê trong đó giả thuyết không được kiểm tra đối với giả thuyết thay thế theo cả hai hướng, tức là cả nhỏ hơn hoặc lớn hơn giá trị kỳ vọng.

Kiểm định hai phía (Two-Tailed Test):Giả thuyết thay thế là 𝐻1≠𝜇0



Ví dụ:

Để so sánh chiều cao trung bình của nam và nữ, kiểm định hai phía sẽ kiểm tra liệu có sự khác biệt đáng kể nào về chiều cao giữa hai nhóm hay không, mà không chỉ rõ hướng khác biệt (nam cao hơn hay thấp hơn nữ không được dự đoán trước).

Ứng dụng:

- Kiểm định một phía: Sử dụng khi thống kê có giả định rõ ràng về hướng của mối quan hệ hoặc sự khác biệt.
- Kiểm định hai phía: Thường được sử dụng khi hướng của sự khác biệt chưa được biết hoặc muốn tránh khả năng thiên vị về một phía.

Kiểm định hai phía thường được ưa chuộng hơn trong phân tích thống kê vì nó tránh được rủi ro của việc bỏ sót các khác biệt quan trọng theo hướng không dự đoán trước.

## 9. Các phương pháp tính giá trị kiểm định thống kê

Các bước thực hiện như sau:

- Xác định rõ vấn đề cần giải quyết và đối tượng nghiên cứu liên quan. Điều này có thể bao gồm việc xác định một câu hỏi nghiên cứu hoặc giả thuyết cụ thể.

- Chọn phương pháp kiểm định thống kê phù hợp dựa trên loại dữ liệu và câu hỏi nghiên cứu. Một số phương pháp kiểm định phổ biến bao gồm kiểm định t, ANOVA, phân tích hồi quy và kiểm định chi-square.

- Xác định mức ý nghĩa để đưa ra quyết định. Mức ý nghĩa này thường được đặt là 0,05, nghĩa là có 5% khả năng kết quả của bạn xảy ra do ngẫu nhiên.

- Thu thập dữ liệu từ đối tượng nghiên cứu của bạn bằng các phương pháp lấy mẫu phù hợp để đảm bảo mẫu đại diện cho tổng thể.

- Tính toán các thống kê phù hợp dựa trên phương pháp kiểm định đã chọn và dữ liệu thu thập được.

- Diễn giải kết quả phân tích thống kê, xem xét mức ý nghĩa đã đặt và các yếu tố liên quan khác. Xác định liệu kết quả có hỗ trợ hoặc bác bỏ câu hỏi nghiên cứu hoặc giả thuyết của bạn.

- Rút ra kết luận dựa trên phân tích thống kê và trình bày chúng rõ ràng với người khác, đồng thời xem xét bất kỳ hạn chế hoặc giả định nào có thể ảnh hưởng đến kết quả của bạn.

### 9.1. Sử dụng Critical Value

- Xác định giả thuyết không và giả thuyết thay thế.

- Chọn thống kê kiểm định phù hợp và tính giá trị của nó dựa trên dữ liệu mẫu.

- Xác định mức ý nghĩa (α) cho kiểm định.

- Tra cứu giá trị tới hạn tương ứng với mức ý nghĩa đã chọn và bậc tự do của thống kê kiểm định.

- So sánh giá trị thống kê kiểm định đã tính với giá trị tới hạn.

- Nếu giá trị thống kê kiểm định đã tính lớn hơn giá trị tới hạn, bác bỏ giả thuyết không.

- Nếu giá trị thống kê kiểm định nhỏ hơn hoặc bằng giá trị tới hạn, không bác bỏ giả thuyết không

Ví dụ, cần kiểm tra xem trọng lượng trung bình của táo có lớn hơn 50g hay không. Lấy mẫu 30 quả táo và tính toán trọng lượng trung bình của mẫu là 52g với độ lệch chuẩn là 4g. Tính giá trị tới hạn với mức ý nghĩa là 0.05 để kiểm tra giả thuyết null rằng trọng lượng trung bình là < 50g.

Đặt giả thuyết:
- H₀: Trọng lượng trung bình của táo <= 50g.
- H₁: Trọng lượng trung bình của táo > 50g.

Vì ta đang kiểm tra xem trọng lượng trung bình có lớn hơn 50 hay không, đây là bài kiểm định một phía (upper tail test), vì vậy vùng tới hạn sẽ nằm ở phía bên phải với α=0.05. Điều này có nghĩa là diện tích đến giá trị tới hạn trên sẽ là 1−0.05=0.95. Do đó, cần tìm giá trị z tương ứng với xác suất 0.95. Từ bảng z, giá trị z critical (z-crit) sẽ là 1.65

Công thức tính giá trị tới hạn là:

    CV=μ ± Zcrit × σ/sqrt(n)
 
Vì đây là kiểm định một phía, nên thay vì sử dụng ký hiệu ±, chỉ sử dụng dấu - để tính giá trị tới hạn phía dưới.

Gía trị tới hạn (CV) = 50 − 1.65 × (4/sqrt(30)) = 48.8

Vì giá trị tới hạn tính được là 48.8, nhỏ hơn giá trị trung bình mẫu là 52 gram, điều này có nghĩa là giá trị trung bình mẫu nằm trong vùng bác bỏ giả thuyết. Do đó, bác bỏ giả thuyết không.

Giá trị z kiểm định tính được là:

(52−50)/(4/sqrt(30))=2.74

Vì Zcalculated=2.74 lớn hơn giá trị Zcrit=1.65, nên giá trị trung bình mẫu nằm trong vùng bác bỏ giả thuyết không.

Kết luận: Bác bỏ giả thuyết không (H0), trọng lượng trung bình của táo lớn hơn 50g với mức ý nghĩa α=0.05

Phương pháp giá trị tới hạn là một công cụ hữu ích trong quyết định thống kê. Nó cung cấp một ngưỡng rõ ràng để bác bỏ hoặc không bác bỏ giả thuyết không dựa trên thống kê kiểm định và mức ý nghĩa.

### 9.2. Sử dụng P Value

Phương pháp p-value là một kỹ thuật thống kê phổ biến để đưa ra quyết định dựa trên dữ liệu. Trong phương pháp này, một giả thuyết được thiết lập và kiểm tra bằng cách sử dụng một bài kiểm định thống kê. 

P-value được tính toán, thể hiện xác suất thu được kết quả cực đoan hoặc hơn thế so với kết quả quan sát được, với giả định rằng giả thuyết không (H₀) là đúng.

- Nếu p-value nhỏ hơn mức ý nghĩa đã xác định trước (thường là 0.05), giả thuyết không sẽ bị bác bỏ và giả thuyết thay thế (H₁) được chấp nhận.

- Nếu p-value lớn hơn mức ý nghĩa, giả thuyết không không bị bác bỏ, và giả thuyết thay thế không được chấp nhận.

Điều quan trọng cần lưu ý là việc bác bỏ giả thuyết không không đồng nghĩa với việc giả thuyết thay thế là đúng, mà chỉ ra rằng dữ liệu cung cấp bằng chứng chống lại giả thuyết không.

- Tính giá trị z-score cho điểm trung bình mẫu trên phân phối:

$$𝑧= \frac{\bar{𝑥}−𝜇}{\frac{\sigma}{\sqrt{n}}}$$

Trong đó:
$\bar{𝑥}$ : Giá trị trung bình của mẫu

μ: Giá trị trung bình của tổng thể (theo giả thuyết không)

σ: Độ lệch chuẩn của tổng thể

n: Kích thước mẫu

`- Tính xác suất tích lũy p cho giá trị z-score:`
Dựa vào giá trị z đã tính, tra bảng z để tìm xác suất tích lũy P(z).

- Với kiểm định một phía trái, xác suất tích lũy P(z) chính là giá trị `p`.

- Với kiểm định một phía phải, tính `p = 1 − P(z)`.

- Với kiểm định hai phía, nhân đôi giá trị p: `p = 2 × min(P(z),1−P(z))`

`- Đưa ra quyết định dựa trên giá trị p:`

So sánh giá trị p với mức ý nghĩa (α):

- `Nếu p≤α: Bác bỏ giả thuyết không (H0)`.

- `Nếu p>α: Không bác bỏ giả thuyết không (H0)`.

Ví dụ: giả thuyết rằng chiều cao trung bình của người dân ở một thị trấn là 170 cm. Một mẫu ngẫu nhiên gồm 49 người từ thị trấn đó được lấy và chiều cao của họ được đo. Kết quả cho thấy chiều cao trung bình của mẫu này là 172 cm và độ lệch chuẩn là 5 cm. Kiểm tra giả thuyết sử dụng mức ý nghĩa là 0.05.

Ở đây, giả thuyết sẽ là:

H0 : Chiều cao trung bình μ=170

H 1 : Chiều cao trung bình μ#170

Công thức tính z-score là:

$𝑧= \frac{\bar{𝑥}−𝜇}{\frac{\sigma}{\sqrt{n}}}$

Trong đó:

$\bar{𝑥}$ = 172, σ=5, n=49

Tính toán giá trị z-score:

𝑧 =2.81



Tiếp theo, tra bảng z để tìm xác suất cho z=2.81, kết quả là 0.9975. Suy ra 1 - p(z)=1-0.9975=0.0025

Vì đây là kiểm định hai phía, ta nhân đôi giá trị này:

0.0025×2=0.005

Vì p-value nhỏ mức ý nghĩa (0.05), ta bác bỏ giả thuyết không.


## 10. Bài tập

1. Một nhà nghiên cứu muốn ước tính trọng lượng trung bình của một loại trái cây nhất định. Một mẫu gồm 50 trái cây được chọn và trọng lượng trung bình được tìm thấy là 250 gram với độ lệch chuẩn là 20 gram. Tính khoảng tin cậy 95% cho trọng lượng trung bình của loại trái cây này.

2. Một nhà sản xuất tuyên bố rằng trọng lượng trung bình của một hộp ngũ cốc là 12 ounce. Một nhóm bảo vệ quyền lợi người tiêu dùng nghi ngờ rằng trọng lượng trung bình thực sự là ít hơn 12 ounce. Một mẫu ngẫu nhiên gồm 36 hộp được chọn và trọng lượng trung bình được tìm thấy là 11.8 ounce với độ lệch chuẩn là 0.6 ounce. Kiểm tra giả thuyết rằng trọng lượng trung bình ít hơn 12 ounce với mức độ ý nghĩa 0.05 sử dụng phương pháp p value.

3. Một công ty tuyên bố rằng sản phẩm mới của họ có tuổi thọ trung bình ít nhất là năm năm. Một mẫu gồm 25 sản phẩm được kiểm tra và tuổi thọ trung bình mẫu được tìm thấy là 4.8 năm với độ lệch chuẩn là 0.7 năm. Kiểm tra giả thuyết rằng tuổi thọ trung bình ít hơn năm năm với mức độ ý nghĩa 0.01 sử dụng phương pháp p value.

4. Giả sử bạn muốn kiểm tra giả thuyết không rằng giá trị trung bình của dân số bằng 50, chống lại giả thuyết thay thế rằng nó lớn hơn 50. Bạn lấy một mẫu ngẫu nhiên có kích thước 25 từ dân số và có được một giá trị trung bình mẫu là 52 và độ lệch chuẩn mẫu là 5. Sử dụng phương pháp critical value với mức độ ý nghĩa 0.05 để kiểm tra giả thuyết.

5. Giả sử bạn muốn kiểm tra giả thuyết không rằng tỷ lệ dân số bằng 0.4, chống lại giả thuyết thay thế rằng nó ít hơn 0.4. Bạn lấy một mẫu ngẫu nhiên có kích thước 100 từ dân số và có được 35 thành công. Sử dụng phương pháp critical value với mức độ ý nghĩa 0.01 để kiểm tra giả thuyết.

6. Một công ty sản xuất một loại sản phẩm nhất định. Một mẫu gồm 50 sản phẩm được lấy và trọng lượng trung bình được tìm thấy là 500 gram với độ lệch chuẩn là 20 gram. Tính sai số chuẩn của trọng lượng trung bình.

7. Một cuộc khảo sát 1000 người trưởng thành được thực hiện để ước tính số giờ trung bình họ dành để xem TV mỗi ngày. Số giờ trung bình được tìm thấy là 3.5 giờ với độ lệch chuẩn là 1.2 giờ. Tính sai số chuẩn của trung bình.

8. Một cuộc khảo sát 500 người được thực hiện để xác định tỷ lệ người ủng hộ một ứng cử viên chính trị cụ thể. Trong số 500 người được khảo sát, 250 người nói rằng họ ủng hộ ứng cử viên này. Tính khoảng tin cậy 99% cho tỷ lệ thực sự của những người ủng hộ ứng cử viên.
