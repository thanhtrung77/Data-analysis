## 1. Giới thiệu

Phân phối mẫu chính xác (Exact Sampling Distribution) là một khái niệm trong thống kê, chỉ phân phối xác suất của một thống kê (như trung bình mẫu, phương sai mẫu, tỷ lệ mẫu, ...) được thu thập từ lấy mẫu lặp lại của cùng một quần thể.

Nói cách khác, phân phối mẫu chính xác mô tả cách mà giá trị của thống kê thay đổi khi chúng ta lấy nhiều mẫu ngẫu nhiên từ cùng một quần thể, với mỗi mẫu có cùng kích thước. Hay nói cách khác, khi ta lặp lại việc lấy mẫu và tính toán thống kê, các giá trị thống kê sẽ khác nhau vì mỗi mẫu là ngẫu nhiên. Phân phối của tất cả các giá trị thống kê này chính là phân phối mẫu chính xác.

Ví dụ, giả sử cần ước tính trọng lượng trung bình của tất cả các con chó trong một thành phố cụ thể. Người ta có thể lấy một mẫu các con chó từ thành phố đó và tính trọng lượng trung bình của mẫu. Sau đó, quá trình này có thể được lặp lại nhiều lần, mỗi lần lấy một mẫu ngẫu nhiên mới của các con chó và tính trọng lượng trung bình của từng mẫu. Phân phối của các giá trị trung bình thu được sẽ là phân phối mẫu chính xác của trung bình mẫu.

Phân phối mẫu chính xác có thể được suy ra một cách toán học bằng cách sử dụng lý thuyết xác suất và các phương pháp thống kê. Chúng cung cấp những hiểu biết giá trị về các đặc tính của các ước lượng thống kê và các kiểm định giả thuyết. Chúng thường được sử dụng để đưa ra suy luận về các tham số tổng thể dựa trên thống kê mẫu, chẳng hạn như ước lượng khoảng tin cậy hoặc thực hiện các kiểm định giả thuyết.

Dưới đây là các phương pháp thống kê chính.

## 2. CHI-SQUARE test

Chi-Square Test (Kiểm định Chi-Square) là một phương pháp thống kê dùng để kiểm tra mối quan hệ giữa các biến phân loại hoặc để so sánh phân phối quan sát với phân phối kỳ vọng, với giả định rằng các biến là độc lập. Đây là một kiểm định phi tham số, không yêu cầu giả định về phân phối chuẩn của dữ liệu.

Mục đích sử dụng
- Kiểm tra tính độc lập (Chi-Square Test of Independence): Đánh giá liệu hai biến phân loại có liên hệ với nhau hay không.
- Kiểm tra mức độ phù hợp (Chi-Square Goodness of Fit Test): Đánh giá liệu phân phối quan sát của dữ liệu có phù hợp với một phân phối mong đợi hay không.

Công thức tính giá trị Chi-Square (𝜒2):

$$𝜒^2 = \sum\frac{(O_i-E_i)^2}{E_i}$$

Oi​: Tần suất quan sát (Observed frequency).

Ei: Tần suất kỳ vọng (Expected frequency).

Ví dụ 1: trong một cuộc thăm dò ý kiến chính trị, trong đó một mẫu gồm 1.000 cử tri được hỏi để chọn giữa hai ứng viên trong một cuộc bầu cử sắp tới. Dữ liệu thu thập được có thể được sắp xếp vào một bảng:



Câu hỏi đặt ra là liệu sự khác biệt giữa các câu trả lời dành cho hai ứng viên có ý nghĩa thống kê hay không. Để xác định điều này, có thể thực hiện kiểm định Chi-square trên dữ liệu trong bảng chéo.

Kiểm định Chi-square sẽ tạo ra một giá trị p-value. Giá trị này cho biết xác suất xảy ra một sự khác biệt giữa hai ứng viên cực đoan như đã quan sát được, với giả định rằng không có sự khác biệt thực sự nào giữa họ trong tổng thể. Nếu p-value nhỏ hơn mức ý nghĩa được xác định trước (ví dụ: 0.05), có thể bác bỏ giả thuyết gốc rằng không có sự khác biệt giữa hai ứng viên. Khi đó, có thể kết luận rằng sự khác biệt giữa hai ứng viên là có ý nghĩa thống kê.

Ví dụ 2: Giả sử có một nghiên cứu, trong đó 100 người tham gia được sử dụng hai loại thuốc khác nhau (A và B) để điều trị một tình trạng y tế cụ thể. Hiệu quả của hai loại thuốc này sẽ được so sánh. Giả thuyết gốc (null hypothesis) cho kiểm định này là: không có sự khác biệt đáng kể giữa hiệu quả của hai loại thuốc.

Người ta ghi nhận số lượng người tham gia cải thiện tình trạng bệnh khi sử dụng mỗi loại thuốc. Theo kỳ vọng, thuốc A và thuốc B có thể mang lại hiệu quả tích cực cho lần lượt 65 người và 35 người trong số 100 người tham gia. Tuy nhiên, trong thử nghiệm, kết quả quan sát được cho thấy thuốc A chữa khỏi cho 40 người, trong khi thuốc B chữa khỏi cho 60 người.



χ² = [-25^2/65] + [25^2/35]
= [625/ 65] + [625 / 35]
= 9.61 + 17.85
= 27.46

Bậc tự do (Degrees of Freedom - DF) cho kiểm định: đối với một bảng 2x2, công thức tính bậc tự do là:

DF = (Số dòng - 1) x (số cột - 1) = 1 x 1 = 1

Để tìm giá trị p-value liên quan đến thống kê Chi-square và bậc tự do, hãy xét mức ý nghĩa là 0.05. Từ bảng phân phối Chi-square với 1 bậc tự do (DF), critical value được xác định là 3.84.



Vì giá trị Chi-square đã tính toán (27.46) lớn hơn giá trị tới hạn, nên bác bỏ giả thuyết gốc (null hypothesis).

Do đó, có thể kết luận rằng có sự khác biệt có ý nghĩa thống kê giữa hai loại thuốc về hiệu quả điều trị tình trạng bệnh.

Ví dụ 3:

Nghiên cứu mối liên hệ giữa giới tính (Nam, Nữ) và sự yêu thích đồ uống (Cà phê, Trà). Thu thập dữ liệu từ 200 người như sau:



Kiểm tra liệu giới tính và sự yêu thích đồ uống có liên hệ với nhau hay không

Xác định giả thuyết
- Giả thuyết gốc (H0): Giới tính và sự yêu thích đồ uống không có mối liên hệ.
- Giả thuyết thay thế (H1): Giới tính và sự yêu thích đồ uống có mối liên hệ.

Tính tần suất kỳ vọng:

Công thức để tính tần suất kỳ vọng cho mỗi ô:

𝐸𝑖𝑗 = (Tổng dòng 𝑖 × Tổng cột 𝑗)/tổng số mẫu

Tần suất kỳ vọng cho ô Nam - Cà phê:
𝐸11 = (Tổng dòng Nam×Tổng cột Cà phê)/tổng số mẫu=100×120/200=60

Tương tự, tính cho các ô khác:

Nam - Trà: 𝐸12=100×80/200=40

Nữ - Cà phê: 𝐸21=100×120/200=60

Nữ - Trà: 𝐸22=100×80200=40

Tính chi-square:

Nam - Cà phê:(70−60)2/60=1.67

Nam - Trà:(30−40)2/40=2.5

Nữ - Cà phê:(50−60)2/60=1.67

Nữ - Trà:(50−40)2/40=2.5

Tổng giá trị Chi-square:

𝜒2=1.67+2.5+1.67+2.5=8.34

Tính số bậc tự do: 

DF = (Số dòng - 1) x (số cột - 1) = 1 x 1 = 1

Tra bảng phân phối Chi-square với 

DF=1 và mức ý nghĩa α=0.05:Giá trị tới hạn = 3.84.

So sánh:

Giá trị Chi-square tính được (8.34) > Giá trị tới hạn (3.84).

Kết luận

Bác bỏ giả thuyết gốc (H0). Có mối liên hệ có ý nghĩa thống kê giữa giới tính và sự yêu thích đồ uống


```python
import numpy as np
from scipy.stats import chi2_contingency

# Bảng tần suất quan sát
data = np.array([[70, 30], [50, 50]])

# Tính Chi-square
chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Kiểm tra mức ý nghĩa
if p < 0.05:
    print("Có mối liên hệ có ý nghĩa thống kê giữa giới tính và sự yêu thích đồ uống.")
else:
    print("Không có mối liên hệ có ý nghĩa thống kê giữa giới tính và sự yêu thích đồ uống.")
```

## 3. Z test

Kiểm định z (z-test) là một phương pháp thống kê được sử dụng để xác định liệu hai trung bình của hai tổng thể có khác biệt hay không, trong trường hợp kích thước mẫu lớn và phương sai của tổng thể đã được biết. Đây là một kiểm định giả thuyết nhằm so sánh trung bình của hai mẫu, thường đến từ các tổng thể có phân phối. 

Kiểm định z được đặt tên theo phân phối chuẩn hóa (standard normal distribution), là một phân phối xác suất với trung bình bằng 0 và độ lệch chuẩn bằng 1. Phương pháp hoạt động bằng cách chuyển đổi các giá trị trung bình mẫu thành z-score, biểu thị số độ lệch chuẩn mà giá trị trung bình mẫu cách trung bình tổng thể. Z-score sau đó được so sánh với một giá trị tới hạn từ phân phối chuẩn hóa để quyết định bác bỏ hay không bác bỏ giả thuyết gốc (null hypothesis).

Trong đó các điều kiện sau được đảm bảo:

- Kích thước mẫu lớn (thường là n>30) hoặc mẫu đến từ tổng thể có phân phối chuẩn.
- Phương sai của tổng thể đã biết (hoặc có thể ước lượng hợp lý).

Các loại Z-test
- One-sample Z-test: Kiểm tra xem trung bình của một mẫu có khác biệt so với một giá trị trung bình giả định (trung bình tổng thể) hay không.

- Two-sample Z-test: Kiểm tra xem hai trung bình mẫu có khác biệt đáng kể hay không, thường để so sánh hai nhóm.

Các bước thực hiện: 

Xác định giả thuyết

- Giả thuyết gốc (H0): Không có sự khác biệt.
- Giả thuyết thay thế (H1): Có sự khác biệt.

Xác định mức ý nghĩa (α)
Thường sử dụng α=0.05.

Tính z-score

Tra bảng phân phối chuẩn hóa
- So sánh giá trị z tính được với giá trị tới hạn zα từ bảng phân phối chuẩn hóa.

- Nếu ∣𝑧∣>zα, bác bỏ H0: Có sự khác biệt có ý nghĩa thống kê.
- Nếu ∣𝑧∣≤zα, không bác bỏ H0: Không có đủ bằng chứng để kết luận có sự khác biệt.

### 3.1. One-sample Z-test

Z-test tính toán giá trị z-score, là số độ lệch chuẩn mà giá trị trung bình mẫu (hoặc khác biệt trung bình mẫu) cách trung bình tổng thể. Công thức cơ bản:

$$𝑧= \frac{\bar{𝑥}−𝜇}{\frac{\sigma}{\sqrt{n}}}$$
​
Trong đó:  

${\bar{𝑥}}$: Trung bình mẫu

μ: Trung bình giả định (hoặc trung bình tổng thể)

σ: Độ lệch chuẩn của tổng thể

n: Kích thước mẫu

Ví dụ: Giả sử chiều cao trung bình của người trưởng thành trong một thành phố là 170cm, với độ lệch chuẩn 10cm. Một mẫu gồm 50 người trưởng thành được chọn ngẫu nhiên, với chiều cao trung bình mẫu là 173cm. Bạn muốn kiểm tra liệu chiều cao trung bình của thành phố có thay đổi hay không ở mức ý nghĩa 0.05

Bước 1: 

Giả thuyết H0 : Trung bình chiều cao = 170

H1 : Trung bình chiều cao ≠170cm

Bước 2: Tính z-score

𝑧 = (173−170)/(10/sqrt(50)=3/1.41≈2.13

Bước 3: So sánh với giá trị tới hạn

Với mức ý nghĩa 0.05 (kiểm định hai phía), giá trị tới hạn từ bảng phân phối chuẩn hóa là 𝑧𝛼/2=1.96



Bước 4: Kết luận

∣z∣=2.13>1.96, bác bỏ H0.

Kết luận: Có sự khác biệt có ý nghĩa thống kê về chiều cao trung bình trong thành phố.

```python
from scipy.stats import norm

# Dữ liệu
mu = 170  # Trung bình tổng thể
sigma = 10  # Độ lệch chuẩn
n = 50  # Kích thước mẫu
x_bar = 173  # Trung bình mẫu

# Tính z-score
z = (x_bar - mu) / (sigma / (n**0.5))

# Giá trị p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Kết quả
print("Z-score:", z)
print("P-value:", p_value)

# Kiểm tra kết luận
alpha = 0.05
if p_value < alpha:
    print("Bác bỏ H0: Sự khác biệt có ý nghĩa thống kê.")
else:
    print("Không bác bỏ H0: Không có sự khác biệt có ý nghĩa thống kê.")
```

### 3.2. Two-sample Z-test

Được sử dụng để kiểm tra xem có sự khác biệt có ý nghĩa thống kê giữa hai trung bình mẫu hay không, khi kích thước mẫu lớn hoặc phương sai tổng thể đã biết. Công thức cơ bản:

$$𝑧= \frac{\bar{𝑥_1}−\bar{𝑥_2}}{\sqrt{\frac{\sigma_1^2}{n1}+{\frac{\sigma_2^2}{n2}}}}$$
​
Trong đó:  

${\bar{𝑥}}$: Trung bình các mẫu

${\sigma^2}$: Phương sai các mẫu

n: Kích thước các mẫu

Ví dụ: Một nhà nghiên cứu muốn so sánh chiều cao trung bình của nam và nữ trong một khu vực. Giả sử các thông tin sau:

Mẫu 1 (Nam): Trung bình x1 =175, kích thước mẫu n1 =50, độ lệch chuẩn σ1 =10.

Mẫu 2 (Nữ): Trung bình x2 =165, kích thước mẫu n2 =40, độ lệch chuẩn σ2 =12.

Mức ý nghĩa: α=0.05.

Bước 1: Tính Z-score

𝑧=(175−165)/sqrt(10^2/50 + 12^2/40)≈4.22

Bước 2: Tra bảng phân phối chuẩn

Với mức ý nghĩa α=0.05 (kiểm định hai phía), giá trị tới hạn là 

zα/2 =1.96.

Bước 3: So sánh Z-score

Z-score tính được (4.22) > Giá trị tới hạn (1.96).

Bước 4: Kết luận

Bác bỏ giả thuyết gốc (H0).

Kết luận: Có sự khác biệt có ý nghĩa thống kê về chiều cao trung bình giữa nam và nữ

```python
import math

# Dữ liệu
x1, x2 = 175, 165  # Trung bình mẫu
n1, n2 = 50, 40  # Kích thước mẫu
sigma1, sigma2 = 10, 12  # Độ lệch chuẩn tổng thể

# Tính Z-score
z = (x1 - x2) / math.sqrt((sigma1**2 / n1) + (sigma2**2 / n2))

# Mức ý nghĩa
alpha = 0.05
z_critical = 1.96  # Giá trị tới hạn cho kiểm định hai phía ở mức ý nghĩa 0.05

# Kết quả
print("Z-score:", z)
if abs(z) > z_critical:
    print("Bác bỏ H0: Có sự khác biệt có ý nghĩa thống kê giữa hai trung bình.")
else:
    print("Không bác bỏ H0: Không có sự khác biệt có ý nghĩa thống kê giữa hai trung bình.")
```

## 4. T test

Kiểm định t (t-test) là một phương pháp thống kê được sử dụng để xác định liệu hai trung bình của hai tổng thể có khác biệt hay không, trong trường hợp kích thước mẫu nhỏ hoặc phương sai tổng thể chưa được biết. Đây là một kiểm định giả thuyết nhằm so sánh trung bình của hai mẫu, thường lấy từ các tổng thể có phân phối chuẩn. Đây là công cụ phổ biến khi:

- Kích thước mẫu nhỏ (n<30).
- Phương sai tổng thể chưa biết, phải ước lượng từ mẫu.

T test được đặt tên theo phân phối t (t-distribution), là một phân phối xác suất tương tự phân phối chuẩn hóa nhưng có phần đuôi dài hơn. Khi kích thước mẫu tăng, phân phối t tiến gần đến phân phối chuẩn.

T test hoạt động bằng cách tính toán giá trị t-statistic, biểu thị mức độ khác biệt giữa trung bình mẫu so với sự biến thiên bên trong các mẫu.

Giả thuyết gốc (H0): Trung bình của hai tổng thể bằng nhau (μ1 = μ2).

Giả thuyết thay thế (H1): Trung bình của hai tổng thể khác nhau (μ1 = μ2).

Các loại T-test

- One-sample T-test: So sánh trung bình của một mẫu với một giá trị giả định (hoặc trung bình tổng thể).
- Independent Two-sample T-test: So sánh trung bình của hai mẫu độc lập để xác định xem chúng có khác biệt đáng kể hay không.
- Paired T-test (Dependent T-test): So sánh trung bình của hai mẫu có liên quan, chẳng hạn trước và sau khi điều trị trong một thí nghiệm.


## 4.1. One sample T test

So sánh trung bình của một mẫu với một giá trị giả định (μ0)

$$𝑧= \frac{\bar{𝑥}−𝜇}{\frac{s}{\sqrt{n}}}$$
​
Trong đó:  

${\bar{𝑥}}$: Trung bình mẫu

μ: Trung bình giả định (hoặc trung bình tổng thể)

σ: Độ lệch chuẩn của mẫu

n: Kích thước mẫu

Ví dụ: Một công ty sản xuất tuyên bố rằng thời gian sử dụng trung bình của một loại pin do họ sản xuất là 300 giờ. Một khách hàng nghi ngờ tuyên bố này và lấy ngẫu nhiên 20 viên pin từ lô hàng để kiểm tra. Kết quả cho thấy thời gian sử dụng trung bình của mẫu là 290 giờ với độ lệch chuẩn là 15 giờ.

Hãy kiểm tra xem thời gian sử dụng trung bình của lô hàng có khác với 300 giờ hay không, với mức ý nghĩa α=0.05.

Giả thuyết gốc (H0): Thời gian sử dụng trung bình của pin là μ=300 giờ.

Giả thuyết thay thế (H1): Thời gian sử dụng trung bình của pin khác 300 giờ.

Công thức T-test:

𝑡=(290−300)/(15/sqrt(20))≈−2.98

Bậc tự do:

df=n−1=20−1=19

Tra bảng phân phối t với df=19 và mức ý nghĩa α=0.05 (two-tailed test), ta được giá trị tới hạn là ±2.093.



Vì −2.98 nằm ngoài khoảng [−2.093,2.093] ta bác bỏ giả thuyết gốc (H0).

Kết luận:Thời gian sử dụng trung bình của pin không bằng 300 giờ. Có sự khác biệt đáng kể về thời gian sử dụng trung bình so với tuyên bố của công ty.

```python
from scipy.stats import ttest_1samp

# Dữ liệu
sample = [290, 285, 295, 300, 288, 289, 292, 287, 290, 286,
          291, 293, 288, 285, 284, 290, 292, 289, 288, 287]

# One-sample T-test
t_stat, p_value = ttest_1samp(sample, 300)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Kiểm tra mức ý nghĩa
alpha = 0.05
if p_value < alpha:
    print("Bác bỏ giả thuyết gốc (H0).")
else:
    print("Không đủ bằng chứng để bác bỏ giả thuyết gốc (H0).")
```

## 4.2. Independent Two-sample T-test

So sánh trung bình của hai mẫu độc lập với công thức tính:

$$𝑧= \frac{\bar{𝑥_1}−\bar{𝑥_2}}{\sqrt{\frac{\sigma_1^2}{n1}+{\frac{\sigma_2^2}{n2}}}}$$
​
Trong đó:  

${\bar{𝑥}}$: Trung bình các mẫu

${\sigma^2}$: Phương sai các mẫu

n: Kích thước các mẫu

Ví dụ: Giả sử một chuyên gia dinh dưỡng muốn kiểm tra liệu một kế hoạch ăn kiêng mới có dẫn đến giảm cân đáng kể cho người tham gia hay không. Chuyên gia này chọn ngẫu nhiên 20 người tham gia và chia họ thành hai nhóm: nhóm kiểm soát và nhóm thử nghiệm. 

Nhóm kiểm soát tuân theo chế độ ăn uống thông thường của họ, trong khi nhóm thử nghiệm tuân theo kế hoạch ăn kiêng mới. Chuyên gia ghi nhận cân nặng của từng người tham gia ở thời điểm bắt đầu và kết thúc nghiên cứu. Dữ liệu được phân phối chuẩn và phương sai của quần thể là không biết.

Giả thuyết gốc là không có sự khác biệt về giảm cân trung bình giữa nhóm kiểm soát và nhóm thử nghiệm. Giả thuyết thay thế là giảm cân trung bình của nhóm kiểm soát ít hơn so với nhóm thử nghiệm.

Trung bình mẫu về giảm cân của nhóm kiểm soát là 2 pound, với độ lệch chuẩn là 1.5 pound.
Trung bình mẫu về giảm cân của nhóm thử nghiệm là 5 pound, với độ lệch chuẩn là 2 pound.

Để kiểm tra giả thuyết, chuyên gia thực hiện t-test độc lập hai mẫu. Thống kê kiểm định được tính như sau:

𝑡=(5−2)/sqrt(2^2/20 + 1.5^2/20)=2.86

Với mức ý nghĩa α=0.05 và bậc tự do df = (𝑛1+𝑛2−2) =18, ta tra bảng phân phối t hoặc sử dụng phần mềm thống kê, giá trị tới hạn cho kiểm định một phía là 1.734.

Vì giá trị t tính được (2.86) lớn hơn giá trị tới hạn (1.734), ta bác bỏ giả thuyết gốc. Kết luận rằng giảm cân trung bình ở nhóm thử nghiệm lớn hơn đáng kể so với nhóm kiểm soát.

Do đó, chuyên gia dinh dưỡng có thể kết luận rằng kế hoạch ăn kiêng mới mang lại hiệu quả giảm cân đáng kể cho người tham gia so với nhóm kiểm soát.

```python
import scipy.stats as stats
import numpy as np

# Dữ liệu đầu vào
mean_control = 2  # Trung bình nhóm kiểm soát
std_control = 1.5  # Độ lệch chuẩn nhóm kiểm soát
n_control = 10  # Số lượng mẫu nhóm kiểm soát

mean_treatment = 5  # Trung bình nhóm thử nghiệm
std_treatment = 2  # Độ lệch chuẩn nhóm thử nghiệm
n_treatment = 10  # Số lượng mẫu nhóm thử nghiệm

# Tính toán thống kê t
t_statistic = (mean_treatment - mean_control) / np.sqrt(
    (std_control**2 / n_control) + (std_treatment**2 / n_treatment)
)

# Bậc tự do (df)
df = n_control + n_treatment - 2

# Mức ý nghĩa (alpha)
alpha = 0.05

# Giá trị tới hạn (critical value) cho kiểm định một phía
critical_value = stats.t.ppf(1 - alpha, df)

# p-value cho kiểm định
p_value = stats.t.sf(t_statistic, df)

# In kết quả
print(f"T-statistic: {t_statistic:.2f}")
print(f"Degrees of freedom: {df}")
print(f"Critical value (one-tailed, alpha={alpha}): {critical_value:.3f}")
print(f"P-value: {p_value:.4f}")

# Kết luận
if t_statistic > critical_value:
    print("Bác bỏ giả thuyết gốc (H0): Kế hoạch ăn kiêng mới hiệu quả hơn.")
else:
    print("Không bác bỏ giả thuyết gốc (H0): Chưa có bằng chứng kế hoạch ăn kiêng mới hiệu quả hơn.")
```

## 5. F test

Kiểm định F (F-test) là một kiểm định thống kê được sử dụng để so sánh phương sai của hai hoặc nhiều quần thể. Đây là một kiểm định giả thuyết nhằm xác định liệu sự biến thiên giữa hai hoặc nhiều nhóm có khác biệt một cách đáng kể hay không. Kiểm định F được đặt tên theo phân phối F, một phân phối xác suất xuất hiện khi so sánh phương sai của hai quần thể có phân phối chuẩn.

Giả thuyết:

Giả thuyết gốc (H0): Phương sai của các quần thể là bằng nhau.

Giả thuyết khác (H1): Ít nhất một phương sai là khác biệt.

Cách tính thống kê kiểm định F:

Được tính bằng cách chia phương sai của một nhóm cho phương sai của nhóm khác.
Nếu có hơn hai nhóm, kiểm định F sử dụng tỷ lệ giữa phương sai của nhóm lớn nhất và nhỏ nhất.

𝐹=Phương sai nhóm lớn nhất/Phương sai nhóm nhỏ nhất

Nếu có nhiều hơn hai nhóm (trong phân tích ANOVA):

𝐹=Phương sai giữa các nhóm (Between-group variance)/Phương sai trong nhóm (Within-group variance)
 
Kết luận:

Nếu giá trị F tính toán lớn hơn giá trị tới hạn, bác bỏ giả thuyết gốc (H0). Kết luận rằng có sự khác biệt đáng kể về phương sai giữa các nhóm.

Nếu giá trị F nhỏ hơn giá trị tới hạn, không bác bỏ giả thuyết gốc. Kết luận rằng không đủ bằng chứng để khẳng định phương sai khác biệt.

Ứng dụng của F test

- Phân tích phương sai (ANOVA):

Kiểm tra xem trung bình của nhiều nhóm có khác nhau không.

Kiểm tra xem phương sai của các nhóm có đồng nhất không.

- Hồi quy: Xác định xem mô hình hồi quy có giải thích tốt biến phụ thuộc hay không.

Ví dụ: Giả sử bạn là nhà nghiên cứu muốn kiểm tra xem hai nhóm sinh viên (nhóm A và nhóm B) có sự khác biệt về mức độ biến thiên điểm số trong một kỳ thi hay không. Dữ liệu về điểm số như sau:

Nhóm A	Nhóm B
  85	78
  88	74
  92	80
  90	76
  87	82

nA =5, nB =5 (số lượng sinh viên trong mỗi nhóm).
Phương sai nhóm A : 8

Phương sai nhóm B : 12

Giả thuyết kiểm định:

H0 : Phương sai của hai nhóm là bằng nhau 

H1 : Phương sai của hai nhóm là khác nhau (

Công thức kiểm định F:
𝐹=Phương sai lớn hơn/Phương sai nhỏ hơn=12/8=1.5

Bậc tự do và mức ý nghĩa:

Bậc tự do của nhóm A: dfA =nA −1 = 4.

Bậc tự do của nhóm B: dfB =nB −1 = 4.

Mức ý nghĩa: α=0.05.

Dựa trên bảng phân phối F, giá trị tới hạn cho F (4,4) tại mức ý nghĩa 0.05 (hai phía) là 6.39.



Kết luận:

Giá trị F tính được: 1.5

Giá trị tới hạn: 6.39

Vì F=1.5<6.39, ta không bác bỏ giả thuyết gốc.

Kết luận: Không có bằng chứng đủ mạnh để kết luận rằng phương sai điểm số giữa hai nhóm là khác nhau.


```python
import scipy.stats as stats

# Dữ liệu
group_A = [85, 88, 92, 90, 87]
group_B = [78, 74, 80, 76, 82]

# Tính phương sai
var_A = stats.variation(group_A) ** 2
var_B = stats.variation(group_B) ** 2

# Thực hiện F-test
f_value = var_B / var_A
df1 = len(group_B) - 1  # bậc tự do nhóm B
df2 = len(group_A) - 1  # bậc tự do nhóm A
p_value = stats.f.sf(f_value, df1, df2) * 2  # hai phía

print(f"F-value: {f_value:.2f}")
print(f"P-value: {p_value:.4f}")

# Kết luận
alpha = 0.05
if p_value < alpha:
    print("Bác bỏ giả thuyết gốc: Phương sai hai nhóm khác nhau.")
else:
    print("Không bác bỏ giả thuyết gốc: Phương sai hai nhóm tương đương.")
```
