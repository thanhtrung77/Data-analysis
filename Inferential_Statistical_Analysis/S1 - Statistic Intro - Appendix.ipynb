```python
import pandas as pd
import numpy as np
import statistics

df = pd.read_csv('insurance.csv')

def descriptive_insurance(data, column):
    insu = {
        'Mean': data[column].mean(),
        'Median': data[column].median(),
        'Mode': data[column].mode()[0],
        'Min': data[column].min(),
        'Max': data[column].max(),
        'Range': data[column].max() - data[column].min(),
        'Std Dev': data[column].std(),
        'Q1': data[column].quantile(0.25),
        'Q3': data[column].quantile(0.75),
        'IQR':data[column].quantile(0.75)-data[column].quantile(0.25),
        'Skewness': data[column].skew(),
        'Kurtosis': data[column].kurtosis()
    }
    return insu

```

```python
# Phương pháp IQR
Q1 = df['charges'].quantile(0.25)
Q3 = df['charges'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Xác định outliers
outliers = df[(df['charges'] < lower_bound) | (df['charges'] > upper_bound)]
print(f"Số lượng outliers: {len(outliers)}")

plt.figure(figsize=(10,6))
sns.boxplot(x=df['charges'])
plt.title('Boxplot của Charges')
plt.show()
```

```python
df['charges_log'] = np.log(df['charges'])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
sns.histplot(data=df, x='charges', ax=ax1)
ax1.set_title('Charges Distribution')
sns.histplot(data=df, x='charges_log', ax=ax2)
ax2.set_title('New charges Distribution')
plt.show()
```

```python
from scipy import stats

# Box-Cox
charges_boxcox, lambda_param = stats.boxcox(df['charges'])
df['charges_boxcox'] = charges_boxcox
print(f"Optimal lambda: {lambda_param:.3f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
sns.histplot(data=df, x='charges', ax=ax1)
ax1.set_title('Charges Distribution')
sns.histplot(data=df, x='charges_boxcox',ax=ax2)
plt.title('New charges Distribution')
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df['charges_robust'] = scaler.fit_transform(df[['charges']])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
sns.histplot(data=df, x='charges', ax=ax1)
ax1.set_title('Charges Distribution')
sns.histplot(data=df, x='charges_robust',ax=ax2)
plt.title('New charges Distribution')
plt.show()
```
