## 1. Correlation

- Correlation (H s tng quan r) c s dng  o lng v nh gi mc  mnh m v hng ca mi quan h gia hai bin s. 
- H s tng quan c gi tr t -1 n 1, cho php xc nh khng ch c mi quan h gia hai bin m cn c tnh cht ca mi quan h .
    - r=1: Tng quan dng hon ho. Hai bin s c mi quan h tuyn tnh tuyt i v cng chiu. Khi mt bin tng, bin kia cng tng theo t l tng ng.
    - r=1: Tng quan m hon ho. Hai bin s c mi quan h tuyn tnh tuyt i nhng ngc chiu. Khi mt bin tng, bin kia gim theo t l tng ng.
    - r=0:Khng c mi quan h tuyn tnh. Hai bin khng lin quan n nhau v mt tuyn tnh, tc l khi mt bin thay i, bin kia khng thay i mt cch c quy lut no.

Thng thng,
 
    - 0<r<1: Tng quan dng. Hai bin c mi quan h cng chiu, nhng khng hon ho. Khi mt bin tng, bin kia c xu hng tng, nhng c th khng theo t l tuyn tnh chnh xc.
    - 1<r<0: Tng quan m. Hai bin c mi quan h ngc chiu, nhng khng hon ho. Khi mt bin tng, bin kia c xu hng gim, nhng khng theo t l chnh xc.



V d, c mt mi tng quan dng gia ht thuc l v nguy c ung th phi. Khi s lng thuc l ht mi ngy tng, nguy c pht trin ung th phi cng tng. 

Ngc li, c mt mi tng quan m gia vic tp th dc v cn nng c th. Khi mc  tp th dc tng, cn nng c th c xu hng gim.


Cc loi h s tng quan:

- H s tng quan Pearson: y l loi h s tng quan ph bin nht, c s dng  o lng mi quan h tuyn tnh gia hai bin s lin tc. N c tnh bng cch chia covariance ca hai bin cho tch ca  lch chun ca chng.
- H s tng quan Spearman: c s dng  o lng mi quan h gia hai bin s khng nht thit phi c phn phi chun, hoc khi bin c th l th bc. H s ny da trn th t ca cc gi tr thay v gi tr thc t.
- H s tng quan Kendall: Tng t nh Spearman, nhng da trn s lng cp m th t ca chng ging nhau v khc nhau gia hai bin.

H s tng quan th hng Spearman o lng mc  m th hng ca hai bin c mi quan h tuyn tnh vi nhau. 

K hiu:  (c gi tr trong khong t -1 n +1)

- -1 biu th mt tng quan m hon ho.
- 0 biu th khng c tng quan.
- +1 biu th mt tng quan dng hon ho.

H s tng quan th hng Spearman da trn s chnh lch gia th hng ca hai bin v c kh nng chng nhiu tt i vi cc gi tr ngoi lai

## 4. Hi quy

Hi quy c s dng  xc nh mi quan h gia mt bin ph thuc v mt hoc nhiu bin c lp.

Mc tiu ca phn tch hi quy l to ra mt m hnh ton hc m t mi quan h gia cc bin. N cng c th c s dng  d on gi tr ca bin ph thuc da trn cc gi tr ca bin c lp.

C nhiu loi phn tch hi quy khc nhau, bao gm hi quy tuyn tnh, hi quy logistic, hi quy a thc,....

Hi quy tuyn tnh l loi phn tch hi quy c s dng ph bin nht, lin quan n vic tm mt ng thng ph hp nht vi cc im d liu trong biu  phn tn.

Phn tch hi quy c th c p dng trong nhiu lnh vc, bao gm kinh t, ti chnh, k thut v khoa hc x hi,...  nghin cu mi quan h gia cc bin v a ra d on v kt qu trong tng lai.

### 4.1 Hi quy tuyn tnh

Hi quy tuyn tnh c s dng  m hnh ha mi quan h gia mt bin ph thuc (thng c k hiu l Y) v mt hoc nhiu bin c lp (thng c k hiu l X). Mc tiu ca hi quy tuyn tnh l tm ra ng thng ph hp nht c th gii thch mi quan h gia cc bin.

C hai loi hi quy tuyn tnh: Hi quy tuyn tnh n bin v Hi quy tuyn tnh a bin.

Hi quy tuyn tnh n bin (Simple Linear Regression): L khi c mt bin c lp v mt bin ph thuc. M hnh ny tm ra ng thng tt nht  m t mi quan h gia hai bin ny.

Hi quy tuyn tnh a bin (Multiple Linear Regression): L khi c nhiu hn mt bin c lp v mt bin ph thuc. M hnh ny tm ra mt mt phng hoc siu phng  m t mi quan h gia cc bin c lp v bin ph thuc.

### a. n bin

Ch c mt bin c lp X. Mi quan h gia X v Y c m hnh ha bng phng trnh ng thng.

$$Y=b0+b1.X$$

Trong ,

b1: l h s  dc (slope).

### 4.4 H s hi quy

H s hi quy l cc gi tr dng s. Chng i din cho mi quan h gia mt bin ph thuc v mt hoc nhiu bin c lp trong m hnh hi quy. Cc h s ny c c lng bng cc phng php thng k, chng hn nh least square error. Chng c s dng  nh lng mc  mnh m v hng ca mi quan h gia cc bin.

Trong m hnh hi quy tuyn tnh n vi mt bin c lp, h s hi quy i din cho s thay i ca bin ph thuc khi bin c lp thay i mt n v. V d, nu h s hi quy l 2,5, th s gia tng mt n v trong bin c lp s i km vi s gia tng 2,5 n v trong bin ph thuc.

Trong m hnh hi quy bi vi hai hoc nhiu bin c lp, mi h s hi quy i din cho s thay i ca bin ph thuc khi bin c lp tng ng thay i mt n v. Tt c cc bin cn li c gi nguyn. Xem xt mt m hnh hi quy bi vi hai bin c lp X1 v X2. H s hi quy 2,5 i vi X1 c ngha l s gia tng mt n v trong X1 s i km vi s gia tng 2,5 n v trong bin ph thuc, trong khi X2 khng i.

H s hi quy rt quan trng trong vic gii thch kt qu phn tch hi quy, v chng cung cp thng tin v hng i v  ln ca mi quan h gia cc bin. Chng cng c th c s dng  d on bin ph thuc da trn cc gi tr ca cc bin c lp.
