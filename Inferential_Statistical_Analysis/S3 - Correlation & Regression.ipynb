## 1. Correlation

- Correlation (Hệ số tương quan r) được sử dụng để đo lường và đánh giá mức độ mạnh mẽ và hướng của mối quan hệ giữa hai biến số. 
- Hệ số tương quan có giá trị từ -1 đến 1, cho phép xác định không chỉ có mối quan hệ giữa hai biến mà còn cả tính chất của mối quan hệ đó.
    - r=1: Tương quan dương hoàn hảo. Hai biến số có mối quan hệ tuyến tính tuyệt đối và cùng chiều. Khi một biến tăng, biến kia cũng tăng theo tỷ lệ tương ứng.
    - r=−1: Tương quan âm hoàn hảo. Hai biến số có mối quan hệ tuyến tính tuyệt đối nhưng ngược chiều. Khi một biến tăng, biến kia giảm theo tỷ lệ tương ứng.
    - r=0:Không có mối quan hệ tuyến tính. Hai biến không liên quan đến nhau về mặt tuyến tính, tức là khi một biến thay đổi, biến kia không thay đổi một cách có quy luật nào.

Thông thường,
 
    - 0<r<1: Tương quan dương. Hai biến có mối quan hệ cùng chiều, nhưng không hoàn hảo. Khi một biến tăng, biến kia có xu hướng tăng, nhưng có thể không theo tỷ lệ tuyến tính chính xác.
    - −1<r<0: Tương quan âm. Hai biến có mối quan hệ ngược chiều, nhưng không hoàn hảo. Khi một biến tăng, biến kia có xu hướng giảm, nhưng không theo tỷ lệ chính xác.



Ví dụ, có một mối tương quan dương giữa hút thuốc lá và nguy cơ ung thư phổi. Khi số lượng thuốc lá hút mỗi ngày tăng, nguy cơ phát triển ung thư phổi cũng tăng. 

Ngược lại, có một mối tương quan âm giữa việc tập thể dục và cân nặng cơ thể. Khi mức độ tập thể dục tăng, cân nặng cơ thể có xu hướng giảm.


Các loại hệ số tương quan:

- Hệ số tương quan Pearson: Đây là loại hệ số tương quan phổ biến nhất, được sử dụng để đo lường mối quan hệ tuyến tính giữa hai biến số liên tục. Nó được tính bằng cách chia covariance của hai biến cho tích của độ lệch chuẩn của chúng.
- Hệ số tương quan Spearman: Được sử dụng để đo lường mối quan hệ giữa hai biến số không nhất thiết phải có phân phối chuẩn, hoặc khi biến có thể là thứ bậc. Hệ số này dựa trên thứ tự của các giá trị thay vì giá trị thực tế.
- Hệ số tương quan Kendall: Tương tự như Spearman, nhưng dựa trên số lượng cặp mà thứ tự của chúng giống nhau và khác nhau giữa hai biến.

## 2. Hệ số tương quan Pearson

$$\text{r} = \frac{{\text{cov(x, y)}}}{\sigma_x.\sigma_y}$$

Trong đó,

$\text{cov(x,y)}$ là gía trị của hiệp phương sai giữa $\text{x,y}$

$\sigma_x,\sigma_y$ là độ lệch chuẩn tương ứng của x và y

Ví dụ:

Tính hệ số tương quan của 2 biến:

X  Y

2  3

3  5

4  4

5  6

6  7

Với

$\text{cov(x, y)} = \frac{\sum_{i=1}^{n} (x_i-\bar{\text{x}})(y_i-\bar{\text{y}})}{n}$

$\sigma_x = \sqrt{\frac{\sum_{i=1}^{n}(x-\bar{x})^2}{n}}$

Trung bình của x (𝒙̅)= (2+3+4+5+6)/5 = 20/5 = 4

Trung bình của y (𝒚̅) = (3+5+4+6+7)/5 = 25/5 = 5



Cov(X,Y) = 9/5 = 1.8

𝝈𝒙 = (10/5)1/2 = 1.414

𝝈𝒚 = (10/5)1/2 = 1.414

r = 1.8/(1.414 * 1.414) = 0.9 #Tương quan dương mạnh

## 3. Hệ số tương quan Spearman (Rank Correlation)

Được sử dụng để đánh giá độ mạnh và hướng của mối quan hệ giữa hai biến bằng cách so sánh thứ hạng của chúng. Có nhiều loại thước đo tương quan thứ hạng, nhưng loại được sử dụng phổ biến nhất là hệ số tương quan thứ hạng Spearman.

Hệ số tương quan thứ hạng Spearman đo lường mức độ mà thứ hạng của hai biến có mối quan hệ tuyến tính với nhau. 

Ký hiệu: ρ (có giá trị trong khoảng từ -1 đến +1)

- -1 biểu thị một tương quan âm hoàn hảo.
- 0 biểu thị không có tương quan.
- +1 biểu thị một tương quan dương hoàn hảo.

Hệ số tương quan thứ hạng Spearman dựa trên sự chênh lệch giữa thứ hạng của hai biến và có khả năng chống nhiễu tốt đối với các giá trị ngoại lai

$$\text{r} = \frac{1-\sum{d^2}}{n.(n^2-1)}$$

Trong đó,

$\sum{d^2}$ là tổng của bình phương các chênh lệch giữa thứ hạng của hai biến, và n là số lượng quan sát

$$\text{r} = \frac{1-6\sum{R^2}}{n.(n^2-1)}$$

$\sum{R^2}$ là tổng của bình phương thứ hạng của các chênh lệch giữa hai biến

Ví dụ:

Tính hệ số tương quan của 2 biến:

X  Y

10 12

7  5

5  8

3  1

6  7

8  10

2  3

4  4

9  11

1  2

Để tính hệ số tương quan thứ hạng Spearman, đầu tiên hãy xếp hạng các giá trị của mỗi biến theo thứ tự từ thấp đến cao. Xếp hạng bằng cách sắp xếp các giá trị và sau đó gán thứ hạng dựa trên vị trí của chúng trong danh sách đã sắp xếp.

Nếu có các giá trị trùng (tức là nhiều giá trị có cùng thứ hạng), hãy gán thứ hạng trung bình cho các giá trị đó.



r = 1-6*14/10(10^2 -1) = -84/990 = -0.08 #Tương quan âm yếu

#### `Probably error`: của hệ số tương quan thứ hạng đo mức độ lỗi hoặc sự không chắc chắn trong ước lượng mẫu của hệ số tương quan thứ hạng.

Probably error được dùng làm thước đo sự biến đổi của hệ số tương quan thứ hạng mà người ta có thể kỳ vọng nếu nghiên cứu được lặp lại nhiều lần.

$$\text{PE} = \frac{0.6745(1-|r|)}{\sqrt{n-2}}$$

Từ ví dụ trên:

Probable error = 0.6745 * (1 - |0.08|) / sqrt(8) = 0.22

Sai số có thể có là 0.22: Điều này có nghĩa là mức độ chênh lệch có thể xảy ra giữa hệ số tương quan mẫu (giá trị quan sát được) và hệ số tương quan thực tế (giá trị chính xác trong tổng thể) là khoảng 0,22.

Nếu ta lặp lại thí nghiệm nhiều lần với các mẫu khác có cùng kích thước, thì trong khoảng 50% số lần, giá trị thực của hệ số tương quan sẽ nằm trong khoảng từ - 0,22 đến + 0,22.

## 4. Hồi quy

Hồi quy được sử dụng để xác định mối quan hệ giữa một biến phụ thuộc và một hoặc nhiều biến độc lập.

Mục tiêu của phân tích hồi quy là tạo ra một mô hình toán học mô tả mối quan hệ giữa các biến. Nó cũng có thể được sử dụng để dự đoán giá trị của biến phụ thuộc dựa trên các giá trị của biến độc lập.

Có nhiều loại phân tích hồi quy khác nhau, bao gồm hồi quy tuyến tính, hồi quy logistic, hồi quy đa thức,....

Hồi quy tuyến tính là loại phân tích hồi quy được sử dụng phổ biến nhất, liên quan đến việc tìm một đường thẳng phù hợp nhất với các điểm dữ liệu trong biểu đồ phân tán.

Phân tích hồi quy có thể được áp dụng trong nhiều lĩnh vực, bao gồm kinh tế, tài chính, kỹ thuật và khoa học xã hội,... để nghiên cứu mối quan hệ giữa các biến và đưa ra dự đoán về kết quả trong tương lai.

### 4.1 Hồi quy tuyến tính

Hồi quy tuyến tính được sử dụng để mô hình hóa mối quan hệ giữa một biến phụ thuộc (thường được ký hiệu là Y) và một hoặc nhiều biến độc lập (thường được ký hiệu là X). Mục tiêu của hồi quy tuyến tính là tìm ra đường thẳng phù hợp nhất có thể giải thích mối quan hệ giữa các biến.

Có hai loại hồi quy tuyến tính: Hồi quy tuyến tính đơn biến và Hồi quy tuyến tính đa biến.

Hồi quy tuyến tính đơn biến (Simple Linear Regression): Là khi có một biến độc lập và một biến phụ thuộc. Mô hình này tìm ra đường thẳng tốt nhất để mô tả mối quan hệ giữa hai biến này.

Hồi quy tuyến tính đa biến (Multiple Linear Regression): Là khi có nhiều hơn một biến độc lập và một biến phụ thuộc. Mô hình này tìm ra một mặt phẳng hoặc siêu phẳng để mô tả mối quan hệ giữa các biến độc lập và biến phụ thuộc.

### a. Đơn biến

Chỉ có một biến độc lập X. Mối quan hệ giữa X và Y được mô hình hóa bằng phương trình đường thẳng.

$$Y=b0+b1.X$$

Trong đó, 

b0: ​là hệ số chặn (intercept)

b1: là hệ số độ dốc (slope).



Ví dụ: Để dự đoán cân nặng của một người dựa trên chiều cao của họ, chiều cao là biến độc lập X, còn cân nặng là biến phụ thuộc Y. Vì vậy, phương trình hồi quy sẽ trở thành:

Cân nặng= 𝑏0 + 𝑏1 × Chiều cao

Trong đó, 

b0: là giá trị cân nặng khi chiều cao bằng 0

b1: là mức độ thay đổi của cân nặng khi chiều cao thay đổi một đơn vị.

Ví dụ: Cho dữ liệu X (age-tuổi) là biến độc lập, Y (lượng Glocose) là biến dự báo của 6 bệnh nhân



b0 = (5486880 – 4968216)/(68160 - 60516) = 518664/7644 = 67.85

b1 = (121176 - 118818)/ (68160 - 60516) = 2358/7644 = 0.30

Biểu thức hồi quy:

Y= 67.85 + 0.30*X

### b. Đa biến

Trong hồi quy tuyến tính đa biến, có hai hoặc nhiều biến độc lập 
𝑋1, 𝑋2,...,𝑋𝑛

Mối quan hệ giữa các biến này và biến phụ thuộc Y được mô hình hóa bằng phương trình sau:

$$𝑌=𝑏0+𝑏1𝑋1+𝑏2𝑋2+...+𝑏𝑛𝑋𝑛$$

Trong đó,

𝑏0, 𝑏1,𝑏2,...,𝑏𝑛: là các hệ số.

Ví dụ: Để dự đoán cân nặng của một người dựa trên chiều cao và độ tuổi của họ, chiều cao và độ tuổi là các biến độc lập 
𝑋1 và 𝑋2, còn cân nặng là biến phụ thuộc 𝑌. Do đó, phương trình hồi quy sẽ trở thành:

Cân nặng=𝑏0+𝑏1×chiều cao+𝑏2×Độ tuổi

Trong đó, 

b0: là giá trị cân nặng khi chiều cao và độ tuổi bằng 0, 

b1: là sự thay đổi của cân nặng khi chiều cao thay đổi một đơn vị, và 

b2: là sự thay đổi của cân nặng khi độ tuổi thay đổi một đơn vị.

### 4.3 Hồi quy không tuyến tính (phi tuyến)

Hồi quy phi tuyến được sử dụng để mô hình hóa mối quan hệ giữa một biến phụ thuộc và một hoặc nhiều biến độc lập mà không phải là tuyến tính.

Hồi quy tuyến tính giả định một mối quan hệ tuyến tính giữa biến phụ thuộc và các biến độc lập. Trong khi đó, các mô hình hồi quy phi tuyến sử dụng một hàm phi tuyến để mô tả mối quan hệ này.

Việc chọn lựa hàm phi tuyến phù hợp để sử dụng phụ thuộc vào tính chất của dữ liệu đang được phân tích và mô hình lý thuyết cơ bản

Trong nhiều lĩnh vực, chẳng hạn như sinh học và kinh tế học, quá trình tăng trưởng của một đối tượng hoặc hệ thống có thể không phải theo một đường thẳng, mà thường có dạng cong, ví dụ như tăng trưởng theo mô hình sigmoid (S-curve). Hồi quy phi tuyến có thể giúp mô hình hóa quá trình này, ví dụ như sự phát triển của quần thể sinh vật hoặc sự tăng trưởng của một công ty.

Trong dược lý học, mối quan hệ giữa liều thuốc và phản ứng của cơ thể (tác dụng của thuốc) cũng có thể là phi tuyến. Ở một mức độ thấp, tác dụng có thể không đáng kể, nhưng khi liều tăng lên, tác dụng sẽ tăng nhanh và sau đó có thể đạt đến điểm bão hòa, không tăng nữa mặc dù liều thuốc tiếp tục tăng. Hồi quy phi tuyến có thể được sử dụng để mô hình hóa mối quan hệ này

Phương trình biểu diễn: 
$𝒚 = 𝒂𝒙^𝟒 + 𝒃𝒙^𝟑 + 𝒄𝒙^𝟐 + 𝒅𝒙^𝟏 + 𝒆$

Ví dụ: Mối quan hệ giữa sản lượng và lao động (Hàm sản xuất Cobb-Douglas)

Trong kinh tế học, hàm sản xuất Cobb-Douglas thường được sử dụng để mô hình hóa mối quan hệ giữa sản lượng (Y) với hai yếu tố đầu vào chính: vốn (K) và lao động (L):

$$𝑌=𝐴⋅𝐾^𝛼⋅𝐿^𝛽$$

Y: Sản lượng đầu ra (GDP hoặc sản lượng sản xuất).

K: Vốn đầu vào.

L: Lao động đầu vào.

A: Hệ số công nghệ.

α,β: Hệ số đàn hồi của vốn và lao động.

Hàm này không tuyến tính theo các tham số và thường được logarit hóa để dễ dàng ước lượng, nhưng bản chất phi tuyến tính của hàm vẫn được bảo toàn.

### 4.4 Hệ số hồi quy

Hệ số hồi quy là các giá trị dạng số. Chúng đại diện cho mối quan hệ giữa một biến phụ thuộc và một hoặc nhiều biến độc lập trong mô hình hồi quy. Các hệ số này được ước lượng bằng các phương pháp thống kê, chẳng hạn như least square error. Chúng được sử dụng để định lượng mức độ mạnh mẽ và hướng của mối quan hệ giữa các biến.

Trong mô hình hồi quy tuyến tính đơn với một biến độc lập, hệ số hồi quy đại diện cho sự thay đổi của biến phụ thuộc khi biến độc lập thay đổi một đơn vị. Ví dụ, nếu hệ số hồi quy là 2,5, thì sự gia tăng một đơn vị trong biến độc lập sẽ đi kèm với sự gia tăng 2,5 đơn vị trong biến phụ thuộc.

Trong mô hình hồi quy bội với hai hoặc nhiều biến độc lập, mỗi hệ số hồi quy đại diện cho sự thay đổi của biến phụ thuộc khi biến độc lập tương ứng thay đổi một đơn vị. Tất cả các biến còn lại được giữ nguyên. Xem xét một mô hình hồi quy bội với hai biến độc lập X1 và X2. Hệ số hồi quy 2,5 đối với X1 có nghĩa là sự gia tăng một đơn vị trong X1 sẽ đi kèm với sự gia tăng 2,5 đơn vị trong biến phụ thuộc, trong khi X2 không đổi.

Hệ số hồi quy rất quan trọng trong việc giải thích kết quả phân tích hồi quy, vì chúng cung cấp thông tin về hướng đi và độ lớn của mối quan hệ giữa các biến. Chúng cũng có thể được sử dụng để dự đoán biến phụ thuộc dựa trên các giá trị của các biến độc lập.
