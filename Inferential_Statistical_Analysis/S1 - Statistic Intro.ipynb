# 1. Giới thiệu

Trong khoa học dữ liệu, thống kê đóng vai trò quan trọng trong việc trích xuất, phân tích, và diễn giải thông tin từ dữ liệu. Nó cung cấp những phương pháp và công cụ để hiểu rõ hơn về dữ liệu, đưa ra quyết định và dự đoán chính xác hơn.

- Thu thập và làm sạch dữ liệu: Trong khoa học dữ liệu, trước khi tiến hành phân tích, dữ liệu cần được thu thập, làm sạch và xử lý. Thống kê cung cấp các phương pháp để xử lý giá trị missing, phát hiện và loại bỏ các giá trị outlier và làm sạch dữ liệu để cải thiện chất lượng phân tích.

- Phân tích mô tả: Thống kê mô tả giúp tóm tắt và trình bày dữ liệu một cách dễ hiểu, sử dụng các thước đo như trung bình, trung vị, mode, phương sai và độ lệch chuẩn. Điều này giúp nhà khoa học dữ liệu hiểu rõ hơn về xu hướng và sự phân bố của dữ liệu.

- Phân tích suy luận: Thống kê suy luận cho phép đưa ra các kết luận hoặc dự đoán về một quần thể (population) dựa trên mẫu (sample) dữ liệu. Các kỹ thuật như kiểm định giả thuyết, khoảng tin cậy, và ước lượng tham số giúp xác định độ tin cậy của các kết luận này.

- Phân tích tương quan và hồi quy: Khoa học dữ liệu thường cần đánh giá mối quan hệ giữa các biến số. Thống kê cung cấp các phương pháp phân tích tương quan và hồi quy để xác định mối liên hệ giữa các biến, giúp dự đoán các kết quả và tối ưu hóa các mô hình dự báo.

- Xác suất: Thống kê sử dụng lý thuyết xác suất để đánh giá mức độ không chắc chắn và tính toán các xác suất của các sự kiện. Điều này rất quan trọng trong các bài toán dự đoán, học máy.

- Kiểm tra mô hình: Sau khi xây dựng mô hình dự đoán, thống kê giúp kiểm tra tính chính xác và độ tin cậy của mô hình đó thông qua các phép kiểm định và chỉ số đánh giá.

Ứng dụng của thống kê:

- Kiểm soát chất lượng: Thống kê thường được sử dụng để đảm bảo chất lượng của sản phẩm hoặc dịch vụ. Các kỹ thuật kiểm soát chất lượng như kiểm soát quá trình bằng thống kê và lấy mẫu có thể giúp xác định các lỗi hoặc biến động trong quy trình sản xuất.

- Nghiên cứu thị trường: Thống kê được sử dụng thường xuyên trong nghiên cứu thị trường để thu thập và phân tích dữ liệu về hành vi và sở thích của người tiêu dùng. Các kỹ thuật như survey và focus group thường được sử dụng để thu thập dữ liệu, và thống kê được dùng để xác định các xu hướng và mô hình trong dữ liệu.

- Y tế: Thống kê đóng vai trò quan trọng trong dịch tễ học, nghiên cứu sự lây lan và kiểm soát bệnh tật. Các nhà dịch tễ học sử dụng kỹ thuật thống kê để phân tích dữ liệu về tỷ lệ mắc của các bệnh, cũng như để xác định các yếu tố rủi ro và xu hướng.

- Tài chính: Thống kê được sử dụng rộng rãi trong tài chính cho quản lý rủi ro và phân tích đầu tư. Các kỹ thuật như phân tích hồi quy, time series analysis, và Monte Carlo thường được sử dụng để phân tích dữ liệu tài chính và đưa ra các quyết định đầu tư.

- Khoa học xã hội: Thống kê được sử dụng nhiều trong khoa học xã hội để phân tích dữ liệu về hành vi con người và các hiện tượng xã hội. Các kỹ thuật như phân tích hồi quy, phân tích tương quan và phân tích nhân tố được sử dụng để xác định các mô hình và mối quan hệ trong dữ liệu.

- Thể thao: Thống kê thường được sử dụng trong thể thao để đánh giá hiệu suất của các vận động viên và đưa ra các quyết định chiến lược. Trong bóng đá, các chỉ số như số lần phòng thủ, sút và tỷ lệ tấn công, phòng thủ được sử dụng để đánh giá hiệu suất của cầu thủ và đưa ra quyết định về việc lựa chọn cầu thủ và chiến lược.

# 2. Descriptive Statistics

Thống kê mô tả liên quan đến phân tích và diễn giải dữ liệu, sử dụng số liệu và biểu đồ để tóm tắt và mô tả các đặc điểm chính của tập dữ liệu. Thống kê mô tả giúp tạo ra một bản tóm lược ngắn gọn về một lượng lớn dữ liệu, giúp xác định các mẫu, xu hướng và mối quan hệ trong dữ liệu.



- Central tendency được sử dụng để mô tả trung tâm của một phân phối. Các thước đo xu hướng trung tâm thường được sử dụng nhất là trung bình, trung vị và mode.

    Ví dụ: một giáo viên muốn tính điểm trung bình của học sinh trong kỳ thi. Trong trường hợp này, trung bình (mean) sẽ là thước đo thích hợp để sử dụng. Trung bình được tính bằng cách cộng tất cả các điểm lại và chia cho tổng số học sinh. Kết quả này cho ra một con số đại diện cho điểm trung bình của cả lớp.

- Dispersion: độ phân tán được sử dụng để mô tả độ rộng phổ của một phân phối bao gồm range, phương sai (variance), độ lệch chuẩn (standard deviation), độ lệch (skewness) và độ nhọn (kurtosis). 

    Ví dụ: Độ lệch của phân phối thu nhập thường được tính toán để xác định các chính sách thuế phù hợp hoặc thiết kế các chương trình xã hội nhằm giảm thiểu bất bình đẳng thu nhập.

## a. Tổng thể (population): 

- Là tập hợp chứa tất cả dữ liệu mà ta quan tâm để đưa ra quyết định

- Là dữ liệu mong muốn có, nhưng khó có khả năng thu thập đủ 

- Thường dùng các chỉ số (parameter) để mô tả population: min, max, mean, median, standard deviation, …

## b. Mẫu (sample): 

- Chứa một phần dữ liệu từ population

- Đó là dữ liệu mà ta có (trong trường hợp lý tưởng nhất là nó có thể đại diện cho tập hợp)

- Dùng thống kê (statistic) để phân tích mẫu

## d. Median

- Trung vị là giá trị nằm ở giữa của tập giá trị khi sắp xếp theo thứ tự tăng dần hoặc giảm dần

- Khác với trung bình, trung vị không nhạy cảm với outlier

- Khi có hai giá trị ở giữa, trung vị là trung bình của hai giá trị đó.

- Trung vị cho biết mức trung tâm của dữ liệu, và thường được sử dụng để đại diện cho "giá trị tiêu biểu" trong tập dữ liệu

- Trung vị có thể khác so với mean trong cùng một tập dữ liệu, và khoảng cách giữa trung vị và mean có thể cho biết về độ lệch của phân phối. Khi trung vị và mean chênh lệch nhau nhiều, điều này thường cho thấy dữ liệu không có phân phối chuẩn và có thể chứa các outliers hoặc bị lệch

## e. Mode

- Mode là giá trị “xuất hiện nhiều nhất” trong một biến số

- Nó có thể được áp dụng cho cả biến số và biến phân loại.

- Sử dụng mode khi:

    * Khi cần tìm giá trị phổ biến nhất, được chọn nhiều nhất trong dữ liệu phân loại.
    * Khi dữ liệu có các giá trị rời rạc và không phải lúc nào mean hoặc median cũng phản ánh đúng ý nghĩa.
    * Khi muốn tránh ảnh hưởng của các giá trị bất thường hoặc tập dữ liệu phân phối bị lệch

```python
statistics.mode(a)
```

```python
import matplotlib.pyplot as plt
import statistics
a = [104,98,90,104,104]

#Tính mean, median và mode
mean_val = statistics.mean(a)
median_val = statistics.median(a)
mode_val = statistics.mode(a)

plt.figure(figsize=(10, 6))
plt.hist(a, bins=7, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_val}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median_val}')
plt.axvline(mode_val, color='blue', linestyle='dashed', linewidth=1.5, label=f'Mode: {mode_val}')

plt.title("Mean, Median, và Mode")
plt.xlabel("Value")
plt.ylabel("Count")
plt.legend()

plt.show()
```

## g. Standard deviation (STD) và Variance

- Độ lệch chuẩn đo lường trung bình khoảng cách của từng giá trị so với giá trị trung bình

- Độ lệch chuẩn càng cao, phân phối càng rộng (và ngược lại)

- Phương sai cho biết mức độ biến động của dữ liệu, nhưng vì phương sai là bình phương nên các giá trị lớn bị phóng đại. Độ lệch chuẩn cung cấp một con số dễ hiểu hơn, cho biết khoảng cách trung bình của các điểm dữ liệu đến trung bình của tập dữ liệu.

- Độ lệch chuẩn thường được sử dụng phổ biến hơn để diễn giải độ phân tán của dữ liệu vì nó có cùng đơn vị với dữ liệu, trong khi phương sai thường được sử dụng trong các phép tính toán khác như hồi quy

- Các bước tính STD: 

```python
# * mean = 100
# * Tính khoảng cách đến mean (104-100), (98-100), (90-100), (104-100), (104-100) = 4, -2, -10, 4, 4
# * Tính bình phương khoảng cách: (4)2, (-2)2, (-10)2, (4)2, (4)2 = 16, 4, 100, 16, 16
# * Tính trung bình các giá trị bình phương này = (16+4+100+16+16)/5 = 30.4 = Variance
# * Căn bậc 2 của variance chính là Standard deviation sqrt(30.4) = 5.51

import math
a = [104,98,90,104,104]
mean_val=statistics.mean(a)
print(mean_val)

#VARIANCE
#a1 = ((104-mean_val)**2 + (98-mean_val)**2 + (90-mean_val)**2 + (104-mean_val)**2 + (104-mean_val)**2)/len(a)
a1 = ((104-mean_val)**2 + (98-mean_val)**2 + (90-mean_val)**2 + (104-mean_val)**2 + (104-mean_val)**2)/(len(a)-1)
print(a1)
#STD
a2= math.sqrt(a1)
print(a2)
```

## g. Skewness và Kurtosis

Skewness đo lường độ lệch của phân phối, tức là mức độ đối xứng của phân phối dữ liệu

- Skewness > 0 (Lệch phải): Phân phối có đuôi kéo dài về phía bên phải. Trong trường hợp này, đa số dữ liệu tập trung ở phía trái của biểu đồ, và giá trị trung bình sẽ lớn hơn trung vị.

- Skewness < 0 (Lệch trái): Phân phối có đuôi kéo dài về phía bên trái. Trong trường hợp này, đa số dữ liệu tập trung ở phía phải của biểu đồ, và giá trị trung bình sẽ nhỏ hơn trung vị.

- Skewness ≈ 0 (Đối xứng): Phân phối không lệch và đối xứng xung quanh trung bình. Ví dụ, phân phối chuẩn (normal distribution) có skewness gần bằng 0

$${skew} = \frac{3*(Mean-Median)}{Standard deviation}$$



```python
a= [10,50,30,20,10,20,70]
mean_val=statistics.mean(a)
print(mean_val)
med_val=statistics.median(a)
print(med_val)
std_val=statistics.stdev(a)
print(std_val)
skew=3*(mean_val-med_val)/std_val # CÂU QUAN TRỌNG TRONG ĐỀ LÝ THUYẾTTHUYẾT
print(skew)
#Skewness = 3*(30-20)/20.70 = 30/20.07 = 1.49
```

```python
import pandas as pd

a = [10, 50, 30, 20, 10, 20, 70]
df_a = pd.Series(a)

skew_val = df_a.skew()
print("Skewness của dữ liệu:", skew_val)
```

Kurtosis đo lường "độ nhọn" hoặc mức độ tập trung của dữ liệu ở phần đỉnh của phân phối.

- Kurtosis > 3 (Leptokurtic): Phân phối có đỉnh sắc hơn so với phân phối chuẩn. Điều này cho thấy nhiều giá trị tập trung ở gần trung tâm và có nhiều ngoại lệ (outliers) ở phần đuôi của phân phối.

- Kurtosis < 3 (Platykurtic): Phân phối có đỉnh phẳng hơn so với phân phối chuẩn. Điều này cho thấy dữ liệu ít tập trung ở trung tâm hơn và ít có giá trị ngoại lệ.

- Kurtosis = 3 (Mesokurtic): Phân phối chuẩn được xem là một chuẩn cơ bản để so sánh. Khi so sánh, người ta thường trừ đi 3 để kurtosis của phân phối chuẩn trở thành 0

    $${Kurtosis} = n*\frac{\sum (x_i-\bar{x})^4}{(\sum (x_i-\bar{x})^2)^2}$$

### Bài 11: Xác định độ lệch

```python
import numpy as np  
from scipy import stats  

# Dữ liệu  
d_10 = [12, 13, 54, 56, 25]  

# Tính độ lệch (skewness)  
skewness = stats.skew(d_10)  

# Xác định loại độ lệch  
if skewness > 0:  
    skew_type = "lệch phải (positive skew)"  
elif skewness < 0:  
    skew_type = "lệch trái (negative skew)"  
else:  
    skew_type = "không lệch (zero skew)"  

# In kết quả  
print("Độ lệch (skewness):", skewness)  
print("Loại độ lệch:", skew_type)
```

### Bài 12: Xác định độ nhọn 

```python
kurtosis = stats.kurtosis(d_10)  

if kurtosis > 3:  
    kurt_type = "leptokurtic"  
elif kurtosis < 3:  
    kurt_type = "platykurtic"  
else:  
    kurt_type = "mesokurtic"  

print("Độ nhọn (kurtosis):", kurtosis, "=>", kurt_type)
```
